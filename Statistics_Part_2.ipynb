#1 What is hypothesis testing in statistics?

   Q1-Answer- Hypothesis testing in statistics is a formal procedure used to determine whether there is enough evidence in a sample of data to infer that a certain condition is true for the entire population.It's a fundamental concept in statistical inference, allowing researchers to make data-driven decisions and draw conclusions about populations based on sample data.

**Formulating Hypotheses:**

**Null Hypothesis (H₀):** This is a statement of no effect, no difference, or no relationship in the population. It's the status quo or the assumption being tested. The goal of hypothesis testing is often to try and disprove or reject the null hypothesis. Examples:

The average height of men and women is the same.
A new drug has no effect on blood pressure.

There is no correlation between study time and exam scores.

   

Alternative Hypothesis (H₁ or H<0xE2><0x82><0x90>):

This is a statement that contradicts the null hypothesis.

It represents what the researcher is trying to find evidence for. It suggests that there is an effect, a difference, or a relationship in the population. Examples:

The average height of men is different from the average height of women.

A new

There is a correlation between study time and exam scores.   

The alternative hypothesis can be one-tailed (directional, e.g., men are taller than women) or two-tailed (non-directional, e.g., men's height is different from women's height).

   

**Setting a Significance Level (α):**

The significance level (alpha, α) is the probability of rejecting the null hypothesis when it is actually true (a Type I error). It's a predetermined threshold set by the researcher. Common values for α are 0.05 (5%) and 0.01 (1%).   

A significance level of 0.05 means there's a 5% risk of concluding there's an effect when there isn't one in the population.   

**Collecting and Analyzing Sample Data:**

A representative sample is taken from the population of interest.   

A relevant test statistic is calculated from the sample data. The choice of test statistic depends on the type of data, the distribution, and the hypotheses being tested (e.g., t-statistic, z-statistic, chi-square statistic, F-statistic).   

**Determining the P-value:**

The p-value is the probability of obtaining test results as extreme as, or more extreme than, the results observed in the sample, assuming the null hypothesis is true.   

A small p-value indicates that the observed data are unlikely if the null hypothesis were true, providing evidence against the null hypothesis.   

**Making a Decision:**

The p-value is compared to the significance level (α).   

If the p-value is less than or equal to α (p ≤ α): The null hypothesis is rejected. This means there is statistically significant evidence to support the alternative hypothesis.

If the p-value is greater than α (p > α): The null hypothesis is not rejected. This does not mean the null hypothesis is true, only that there is not enough evidence from the sample data to reject it.   

In essence, hypothesis testing is a process of:

Stating competing hypotheses about a population parameter.

Gathering evidence (sample data).

Assessing the strength of the evidence against the null hypothesis using a statistical test and calculating a p-value.   

Making a decision based on the p-value and the chosen significance level.

**Important Considerations:**

**Type I Error (False Positive):** Rejecting a true null hypothesis (probability = α).   

**Type II Error (False Negative):** Failing to reject a false null hypothesis (probability = β). The power of a test is 1 - β, which is the probability of correctly rejecting a false null hypothesis.   

**Statistical Significance vs. Practical Significance:** A statistically significant result doesn't necessarily mean the effect is practically important or meaningful in the real world. The context and magnitude of the effect should also be considered.

Assumptions of Tests: Most statistical tests have underlying assumptions about the data (e.g., normality, independence, equal variances). Violating these assumptions can affect the validity of the test results.

#3 Q3. What is the significance level in hypothesis testing, and why is it important?

   Q3-Answer- The
significance level (α) in hypothesis testing is the predetermined probability of rejecting the null hypothesis when it is actually true.In simpler terms, it's the maximum acceptable risk of making a Type I error – a false positive conclusion.

**Probability Threshold:** The significance level acts as a threshold. If the probability of observing your sample data (or more extreme data) assuming the null hypothesis is true (the p-value) falls below this threshold, you reject the null hypothesis.   

**Researcher's Choice:** The researcher sets the significance level before conducting the experiment or analysis. Common values are 0.05 (5%) and 0.01 (1%), but other values can be chosen depending on the context of the study.   

**Risk of Incorrect Rejection:** A significance level of 0.05 means there is a 5% chance that you will incorrectly reject the null hypothesis, even if it is true in the population.   

**Why is the significance level important?**
**The significance level plays a crucial role in hypothesis testing for several reasons:**

Establishes a Decision Rule: It provides a clear and objective criterion for deciding whether to reject the null hypothesis. By setting α, the researcher defines how much evidence is required from the sample data to conclude that the effect observed is statistically significant and not likely due to random chance.


**Controls the Probability of a Type I Error:** The primary importance of the significance level is that it directly controls the probability of committing a Type I error (false positive). Researchers want to minimize the risk of concluding there is a real effect or difference when there isn't one in the population. Choosing a smaller α (e.g., 0.01) reduces this risk.   


**Provides a Standard for Comparison:** The chosen significance level serves as a benchmark against which the p-value is compared. This standardized approach helps researchers and readers evaluate the strength of the evidence against the null hypothesis in a consistent manner.   


**Avoids Bias and P-Hacking:** Setting the significance level before analyzing the data is crucial for maintaining the integrity of the research. Deciding on α after seeing the results (a practice known as p-hacking or data dredging) can lead to biased conclusions and an increased risk of false positives.   


**Reflects the Consequences of Errors:** The choice of significance level should consider the potential consequences of making a Type I error in the specific context of the research. For example, in medical research where a false positive could lead to unnecessary treatment or harm, a lower significance level (e.g., 0.01) might be preferred over social science research where the consequences of a false positive might be less severe.

#4 What does a P-value represent in hypothesis testing?

   Q4-Answer- The P-value in hypothesis testing represents the probability of obtaining test results at least as extreme as the results observed in your sample, assuming that the null hypothesis is true.

**Null Hypothesis:** This is a statement of no effect or no difference. It's the default assumption we try to disprove.   

**Observed Data:** These are the results you collected in your study or experiment.

**Extreme Results:** This refers to results that are as far away from what the null hypothesis predicts (or even further away) as the results you actually observed. The definition of "extreme" depends on the direction of your alternative hypothesis (one-tailed or two-tailed test).   

**Probability:** The P-value is a probability, ranging from 0 to 1.   

In simpler terms, the P-value tells you how likely it is that you would have seen your data (or more unusual data) if there was really no effect or difference in the population you're studying.   

**How to interpret a P-value:**

**Small P-value (typically ≤ 0.05):** This indicates that your observed data are unlikely to have occurred if the null hypothesis were true. Therefore, you have evidence to reject the null hypothesis in favor of the alternative hypothesis. The results are considered statistically significant.   

**Large P-value (typically > 0.05):** This indicates that your observed data are reasonably likely to have occurred even if the null hypothesis were true. Therefore, you fail to reject the null hypothesis. This does not mean that the null hypothesis is true, only that there isn't enough evidence to disprove it based on your data. The results are considered not statistically significant.

**Important Considerations:**

The P-value is NOT the probability that the null hypothesis is true. It's the probability of the observed data given that the null hypothesis is true.   

Statistical significance does not equal practical significance. A very small P-value might be obtained with a very large sample size even if the actual effect size is small and not meaningful in the real world.   

The chosen significance level (alpha, often set at 0.05) is a threshold you set before conducting the test. You compare the P-value to this threshold to make your decision about the null hypothesis.   

The P-value is influenced by the sample size. Larger samples tend to produce smaller P-values.

   

#5 How do you interpret the P-value in hypothesis testing?

   Q5-Answer- Interpreting the P-value in hypothesis testing is crucial for drawing meaningful conclusions from your data.

**Compare the P-value to the Significance Level (Alpha):**

Before conducting your hypothesis test, you should have set a significance level (alpha, denoted as α). This is the probability threshold you're willing to accept for rejecting the null hypothesis when it is actually true (a Type I error). Common values for alpha are 0.05, 0.01, and 0.10.   

The core of interpreting the P-value involves comparing it to this predetermined alpha level.   

**Decision Rules:**

**If the P-value is less than or equal to the significance level (P-value≤α):**

You reject the null hypothesis (H0​).   

This means that the observed data provide enough statistical evidence to conclude that the null hypothesis is likely false.   

The results are considered statistically significant at the chosen alpha level.   

You would then typically support the alternative hypothesis (Ha​).   



**If the P-value is greater than the significance level (P-value>α):**

You fail to reject the null hypothesis (H0​).   

This means that the observed data do not provide enough statistical evidence to conclude that the null hypothesis is false at the chosen alpha level.   
The results are considered not statistically significant.   

It's important to note that failing to reject the null hypothesis does not mean that the null hypothesis is true. It simply means that there isn't enough evidence to disprove it based on your data.   



**Understanding the Meaning of the P-value:**

The P-value is the probability of observing data as extreme as, or more extreme than, the data you collected, assuming that the null hypothesis is true.   

**Analogy:**
Imagine a courtroom trial.

**Null Hypothesis (H0​):** The defendant is innocent.


**Alternative Hypothesis (Ha​):** The defendant is guilty.


**Evidence:** The data you collected.


**P-value:** The probability of observing the evidence you have (or even stronger evidence of guilt) if the defendant were actually innocent.

**Small P-value:** If the P-value is small (e.g., less than your chosen alpha), it means that the evidence is unlikely to have occurred if the defendant were innocent. This leads you to reject the assumption of innocence (reject the null hypothesis) and conclude guilt (support the alternative hypothesis).   

**Large P-value:** If the P-value is large, it means that the evidence you have is reasonably likely to have occurred even if the defendant were innocent. You don't have enough strong evidence to overturn the assumption of innocence (fail to reject the null hypothesis). This doesn't mean the defendant is definitely innocent, just that the evidence isn't strong enough to prove guilt.


**Important Caveats:**

Statistical Significance vs. Practical Significance: A statistically significant result (small P-value) doesn't necessarily mean the effect is practically important or meaningful in the real world. The size of the effect also needs to be considered.

The P-value is not the probability that the null hypothesis is true. This is a common misconception. The P-value is conditional on the null hypothesis being true.

The P-value is influenced by sample size. Larger sample sizes can lead to smaller P-values even for small effect sizes.   

**Avoid "Accepting" the Null Hypothesis:** You can only "fail to reject" the null hypothesis. You haven't proven it to be true.

#6 What are Type 1 and Type 2 errors in hypothesis testing?

   Q6-Answer- In hypothesis testing, we aim to determine whether there is enough evidence to reject the null hypothesis (H0​). However, our decision is based on sample data, so there's always a chance of making an incorrect conclusion. These incorrect conclusions are known as Type 1 and Type 2 errors.

**Type 1 Error (False Positive):**


**Definition:** A Type 1 error occurs when you reject the null hypothesis when it is actually true.   
In simpler terms: You conclude that there is a significant effect or difference when, in reality, there isn't one in the population. You've found a "false positive" result.   

**Probability:** The probability of making a Type 1 error is denoted by α (alpha), which is also the significance level you set for your hypothesis test (commonly 0.05).   

**Example:** Imagine a medical test where the null hypothesis is that a patient does not have a certain disease. A Type 1 error would be concluding that the patient has the disease when they actually do not (a false positive diagnosis).   

**Consequences:** The consequences of a Type 1 error can range from unnecessary actions or further investigations to more serious issues depending on the context.   

**Type 2 Error (False Negative):**

**Definition:** A Type 2 error occurs when you fail to reject the null hypothesis when it is actually false.   
In simpler terms: You conclude that there is no significant effect or difference when, in reality, there is one in the population. You've missed a real effect, resulting in a "false negative" result.   

**Probability:** The probability of making a Type 2 error is denoted by β (beta).   

**Power of the Test:** The power of a statistical test is the probability of correctly rejecting a false null hypothesis, and it is calculated as 1−β. A higher power means a lower chance of committing a Type 2 error.   

**Example:** Using the same medical test, a Type 2 error would be concluding that the patient does not have the disease when they actually do (a false negative diagnosis).   

**Consequences:** The consequences of a Type 2 error can also be significant, such as failing to implement an effective treatment or missing a crucial discovery.   

**Analogy to a Criminal Trial:**

**Null Hypothesis (H0​):** The defendant is innocent.


**Alternative Hypothesis (Ha​):** The defendant is guilty.


**Type 1 Error:** Rejecting the null hypothesis when it's true means finding an innocent person guilty.

**Type 2 Error:** Failing to reject the null hypothesis when it's false means letting a guilty person go free.

#7 What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

   Q7-Answer- The key difference between a one-tailed and a two-tailed test in hypothesis testing lies in the directionality of the alternative hypothesis and consequently, the location of the critical region(s) in the probability distribution of the test statistic.

**Alternative Hypothesis (Ha​ or H1​):**

**Two-Tailed Test:** The alternative hypothesis states that the population parameter is not equal to the value specified in the null hypothesis (H0​). It doesn't specify whether the parameter is greater or smaller; it simply asserts a difference.

**Example:** H0​:μ=50 (The population mean is 50)

**Example:** Ha​:μ=50 (The population mean is not 50 - it could be greater or smaller)


**One-Tailed Test:** The alternative hypothesis specifies that the population parameter is either greater than or less than the value specified in the null hypothesis (H0​), but not both. It indicates a directional expectation.   

**Right-Tailed Test:** Ha​:μ>50 (The population mean is greater than 50)

**Left-Tailed Test:** Ha​:μ<50 (The population mean is less than 50)



**Critical Region:**
The critical region is the area in the tail(s) of the sampling distribution where, if the test statistic falls within it, you reject the null hypothesis. The size of this region is determined by the significance level (α).   

**Two-Tailed Test:** The critical region is split into two equal parts, one in each tail of the distribution. For a significance level of α, each tail contains an area of α/2.

You would reject the null hypothesis if your test statistic falls in either the upper or the lower tail.   


**One-Tailed Test:** The entire critical region (with an area of α) is located in only one tail of the distribution.

**Right-Tailed Test:** The critical region is in the upper (right) tail. You reject H0​ if the test statistic is significantly large.

**Left-Tailed Test:** The critical region is in the lower (left) tail. You reject H0​ if the test statistic is significantly small (negative).


**When to Use Which Test:**

**Two-Tailed Test:** Use a two-tailed test when you are interested in detecting a difference in either direction.

You don't have a prior expectation of whether the parameter will be greater or smaller than the null value.

This is the more common and conservative approach.


**One-Tailed Test:** Use a one-tailed test only when you have a strong directional hypothesis based on prior research, theory, or practical considerations. You are only interested in detecting an effect in one specific direction, and an effect in the opposite direction would not be considered meaningful. Using a one-tailed test can have more statistical power to detect an effect in the specified direction because the entire significance level is concentrated in one tail. However, if an effect occurs in the opposite direction (even if it's large), you would not be able to reject the null hypothesis.   


**Example:**
Let's say a company claims its new light bulb has an average lifespan of 60 days.

**Two-Tailed Test:**

**H0​:μ=60**

**Ha​:μ=60 (We want to see if the lifespan is different from 60 days - it could be longer or shorter)**

The critical region would be in both the left and right tails of the distribution.



**One-Tailed Test (Right-Tailed):**

**H0​:μ≤60 (or μ=60)**

**Ha​:μ>60 (We specifically want to see if the lifespan is greater than 60 days)**

The critical region would be only in the right tail of the distribution.


**One-Tailed Test (Left-Tailed):**

**H0​:μ≥60 (or μ=60)**

**Ha​:μ<60 (We specifically want to see if the lifespan is less than 60 days)**

The critical region would be only in the left tail of the distribution.

#8 What is the Z-test, and when is it used in hypothesis testing?

   Q8-Answer- The Z-test is a statistical hypothesis test used to determine if there is a significant difference between the means of two groups or a sample mean and a known population mean. It relies on the assumption that the data is normally distributed (or the sample size is large enough for the Central Limit Theorem to apply) and that the population standard deviation is known.   

When is the Z-test used in hypothesis testing?

**The Z-test is typically used under the following conditions:**


 **Population Standard Deviation is Known (σ):** This is the primary condition. If you know the standard deviation of the entire population you are studying, you can use a Z-test.   


**Large Sample Size (n ≥ 30):** Even if the population standard deviation is unknown, if your sample size is large (generally n ≥ 30), the Central Limit Theorem states that the sampling distribution of the sample mean will approximate a normal distribution. In such cases, the sample standard deviation (s) can be used as an estimate of the population standard deviation (σ), and a Z-test can be used as an approximation. However, many statisticians prefer to use a t-test in this scenario as it accounts for the uncertainty introduced by estimating the population standard deviation.   


 **Data Follows a Normal Distribution:** The Z-test assumes that the data being tested comes from a normally distributed population. While the Central Limit Theorem helps with large samples, for smaller samples, this assumption is more critical.   


**Specific Scenarios for Using a Z-test:**

**One-Sample Z-test:** Used to compare the mean of a single sample to a known or hypothesized population mean when the population standard deviation is known.   

**Example:** Testing if the average height of students in a large university differs from the national average height (assuming the national standard deviation is known).



**Two-Sample Z-test (Independent Samples):** Used to compare the means of two independent groups when the population standard deviations for both groups are known.

**Example:** Comparing the average scores of students taught by two different teaching methods, assuming the standard deviations of scores for both methods are known.


**Paired Z-test (Dependent Samples):** Used to compare the means of two related groups (e.g., before and after measurements on the same individuals) when the population standard deviation of the differences is known.


**Z-test for Proportions:** Used to test hypotheses about population proportions, especially when dealing with large sample sizes.


**Formula for the One-Sample Z-test:**

Z=  (x̄ - μ) / (σ / √n)

x̄ is the sample mean
μ0​ is the hypothesized population mean (from the null hypothesis).   
σ is the population standard deviation.   
n is the sample size.

It essentially standardizes the difference between the observed data and what is expected under the null hypothesis.   


**Location on the Standard Normal Distribution:** When the population is normally distributed or the sample size is large enough (due to the Central Limit Theorem), the Z-score follows a standard normal distribution. This distribution has a mean of 0 and a standard deviation of 1.

**Decision Making:** The Z-score, in conjunction with the critical value or P-value, helps you make a decision about the null hypothesis:


**Reject H0​:** If the Z-score is sufficiently far from zero (falls in the critical region or has a small P-value), you have evidence to reject the null hypothesis in favor of the alternative hypothesis.   

**Fail to Reject H0​:** If the Z-score is close to zero (does not fall in the critical region or has a large P-value), you do not have enough evidence to reject the null hypothesis. This does not mean the null hypothesis is true, only that the data does not provide sufficient evidence against it.

#10 What is the T-distribution, and when should it be used instead of the normal distribution?

   Q10-Answer- The T-distribution, also known as Student's t-distribution, is a probability distribution that is used to estimate population parameters when the sample size is small or when the population standard deviation is unknown. It is similar in shape to the normal distribution (bell-shaped and symmetrical) but has heavier tails. This means that data points are more likely to fall further away from the mean compared to a normal distribution, especially with smaller sample sizes.   

The shape of the t-distribution depends on a parameter called the degrees of freedom (df). For a single sample t-test, the degrees of freedom are typically n−1, where n is the sample size.   

When should the T-distribution be used instead of the normal distribution?

**The t-distribution should be used instead of the normal distribution primarily in the following situations:**


 **Population Standard Deviation is Unknown (σ):** In most real-world scenarios, the population standard deviation is not known. Instead, we estimate it from the sample standard deviation (s). When you are using the sample standard deviation to estimate the population standard deviation, the t-distribution provides a more accurate probability distribution for the sample mean.   


Small Sample Size (n < 30): When the sample size is small (generally considered less than 30), the sample standard deviation is a less reliable estimate of the population standard deviation. The t-distribution accounts for this increased uncertainty by having heavier tails, which leads to wider confidence intervals and more conservative hypothesis tests.   


**Here's a more detailed breakdown of when to use each distribution:**

**Use the Z-distribution (Standard Normal Distribution) when:**

The population standard deviation (σ) is known.

The sample size (n) is large (typically n≥30), even if σ is unknown (due to the Central Limit Theorem, the sample mean will be approximately normally distributed, and the sample standard deviation s provides a good estimate of σ).



**Use the T-distribution (Student's t-distribution) when:**

The population standard deviation (σ) is unknown, and you are using the sample standard deviation (s) in its place.

The sample size (n) is small (typically n<30).

The population from which the sample is drawn is approximately normally distributed (this assumption is more critical for small samples).


**Why the T-distribution is more appropriate in these cases:**

**Accounting for Uncertainty:** When the population standard deviation is unknown and estimated by the sample standard deviation, there's an added layer of uncertainty. The t-distribution's heavier tails reflect this increased variability.   

**Small Sample Reliability:** With small samples, the sample mean might not be as precise an estimate of the population mean as with large samples. The t-distribution's wider shape acknowledges this and provides more realistic probability estimates.

As the sample size (and thus the degrees of freedom) increases, the t-distribution approaches the standard normal distribution. When the degrees of freedom are large (e.g., > 30), the t-distribution and the normal distribution become very similar, and using either will often yield similar results. However, it's generally more statistically correct to use the t-distribution when the population standard deviation is unknown, regardless of sample size, although the difference becomes negligible for large samples.

The heavier tails of the t-distribution account for the increased uncertainty when estimating the population standard deviation from a small sample.   

**How the T-test is Used in Hypothesis Testing:**

**The T-test is used in hypothesis testing through a series of steps:**


**State the Null and Alternative Hypotheses:**

The null hypothesis (H0​) typically states that there is no significant difference between the means being compared.   

The alternative hypothesis (Ha​ or H1​) states that there is a significant difference. This difference can be non-directional (two-tailed test) or directional (one-tailed test).   



**Choose a Significance Level (α):**

This is the probability of rejecting the null hypothesis when it is actually true (Type I error). Common values are 0.05, 0.01, or 0.10.   



**Calculate the Test Statistic (t-value):**

**The formula for the t-statistic varies depending on the type of t-test being used:**

**One-Sample T-test:** Used to compare the mean of a single sample to a known or hypothesized population mean (μ0​) when the population standard deviation is unknown.

t = (x̄ - μ) / (s / √n)

where xˉ is the sample mean, s is the sample standard deviation, and n is the sample size.   

**Independent Two-Sample T-test:** Used to compare the means of two independent groups. There are versions for equal and unequal variances.
Equal Variances (Pooled T-test):

t = (x̄₁ - x̄₂) / √(sₚ²(1/n₁ + 1/n₂))

**where sp​ is the pooled standard deviation.Unequal Variances (Welch's T-test):**

t = (x̄₁ - x̄₂) / √(s₁²/n₁ + s₂²/n₂)

**Paired Sample T-test:**Used to compare the means of two related groups (e.g., before and after measurements on the same subjects).

t = d̄ / SE(d̄), SE(d̄) = sd / √n, and df = n - 1.

where dˉ is the mean of the differences between the paired observations, and sd​ is the standard deviation of the differences.   

   



**Determine the Degrees of Freedom (df):**

**The degrees of freedom depend on the sample size(s) and the type of t-test:**

**One-Sample:** df=n−1   

**Independent Two-Sample (Equal Variances):** df=n1​+n2​−2   

**Independent Two-Sample (Unequal Variances):** The calculation is more complex (Welch-Satterthwaite equation).

**Paired Sample:** df=n−1 (where n is the number of pairs).   

   



**Find the Critical Value or P-value:**

Critical Value Approach: Using the chosen significance level (α) and the degrees of freedom, find the critical t-value(s) from a t-distribution table. The rejection region(s) are determined by the alternative hypothesis (one-tailed or two-tailed). If the calculated t-statistic falls within the rejection region(s) (i.e., its absolute value is greater than the critical value), you reject the null hypothesis.   

**P-value Approach:** Calculate the p-value, which is the probability of obtaining a t-statistic as extreme as, or more extreme than, the one calculated, assuming the null hypothesis is true. This is done using statistical software or t-distribution tables. If the p-value is less than or equal to the significance level (α), you reject the null hypothesis.   



**Make a Decision:**

If you reject the null hypothesis, you conclude that there is statistically significant evidence to support the alternative hypothesis.

If you fail to reject the null hypothesis, you conclude that there is not enough statistically significant evidence to support the alternative hypothesis based on your data.

#13 What is the relationship between Z-test and T-test in hypothesis testing?

   Q13-Answer- The Z-test and the T-test are both fundamental tools in hypothesis testing, used to assess whether there is a statistically significant difference between sample means and population means (or between the means of two samples). Their relationship lies in the fact that the T-test can be considered a more general form that is used when the conditions for a Z-test (primarily knowing the population standard deviation) are not met, especially with smaller sample sizes.

**When to Use Which:**

**Z-test:** Used when the population standard deviation (σ) is known, and the sample size is typically large (n≥30). With a large sample size, even if σ is unknown, the Central Limit Theorem allows us to approximate the sampling distribution as normal, and the sample standard deviation (s) can be a reasonable estimate of σ. However, the Z-test strictly assumes a known σ.   

**T-test:** Used when the population standard deviation (σ) is unknown, and we estimate it using the sample standard deviation (s). It is particularly important when the sample size is small (n<30), but it is also the appropriate test even for large sample sizes when σ is unknown.   

**Test Statistic and Distribution:**

**Z-test:** The test statistic follows a
standard normal distribution (Z-distribution). The formula for a one-sample Z-test is:

z = (x̄ - μ) / (σ / √n)

**T-test:** The test statistic follows a Student's t-distribution (T-distribution). The formula for a one-sample T-test is:

t = (x̄ - μ) / (s / √n)

The t-distribution has heavier tails compared to the standard normal distribution, especially with smaller degrees of freedom (df=n−1), reflecting the increased uncertainty due to estimating σ.   

**Convergence:**

As the sample size (n) increases, the degrees of freedom for the T-test (df=n−1) also increase. As the degrees of freedom grow larger (generally above 30), the t-distribution approaches the standard normal distribution. In such cases, the critical values and p-values obtained from a T-test become very similar to those from a Z-test, and the conclusions of the hypothesis tests are often the same.   

**Assumptions:**
Both tests assume that the data are drawn from a population that is approximately normally distributed. However, the Z-test relies on knowing σ, while the T-test estimates it with s. The T-test is more robust to violations of the normality assumption, especially with larger sample sizes.

#14 What is a confidence interval, and how is it used to interpret statistical results?

   Q14-Answer- A confidence interval (CI) is a range of values, derived from sample data, that is likely to contain the true value of an unknown population parameter. It provides a measure of the uncertainty associated with a sample estimate. Instead of just providing a single point estimate (like the sample mean), a confidence interval gives a range within which the population parameter is likely to fall, with a certain degree of confidence.   

**Key Components of a Confidence Interval:**

**Point Estimate:** This is the sample statistic (e.g., sample mean, sample proportion) used to estimate the population parameter. It's the center of the confidence interval.   

Margin of Error: This is the range of values above and below the point estimate. It accounts for the sampling variability and the desired level of confidence. The margin of error is calculated by multiplying a critical value (from a Z or T distribution) by the standard error of the statistic.   
Confidence Level: This is the probability (expressed as a percentage or decimal) that the confidence interval will contain the true population parameter if we were to take many random samples and construct a confidence interval for each sample. Common confidence levels are 90%, 95%, and 99%.   

Formula for a Basic Confidence Interval (for the mean, when population standard deviation is known or n is large):

Sample Mean ± (Critical Value * Standard Error)

Where:

xˉ is the sample mean.   

Z2α​​ is the critical Z-value corresponding to the chosen confidence level (e.g., 1.96 for 95% confidence).   

σ is the population standard deviation (if known).   

n is the sample size.   

If the population standard deviation is unknown and the sample size is small, a T-distribution and the sample standard deviation (s) are used instead of the Z-distribution and σ.   

**How Confidence Intervals are Used to Interpret Statistical Results:**

Confidence intervals provide a more informative way to interpret statistical results compared to just looking at point estimates or

**P-values alone. Here's how:**

**Estimating the Range of the Population Parameter:** The primary use of a confidence interval is to provide a plausible range of values for the true population parameter. For example, a 95% confidence interval for the average height of adult males might be [175 cm, 180 cm]. This suggests that we are 95% confident that the true average height of all adult males falls within this range.


**Quantifying Uncertainty:** The width of the confidence interval reflects the precision of our estimate. A narrow interval indicates a more precise estimate with less uncertainty, while a wide interval suggests a less precise estimate with more uncertainty. The width is influenced by factors like sample size and the variability of the data.   


**Assessing Statistical Significance (Indirectly):** While P-values are the direct measure of statistical significance in hypothesis testing, confidence intervals can provide related insights.   

If a confidence interval for a difference between two means (or other parameters) does not contain zero, this suggests that the difference is statistically significant at the corresponding alpha level (where α=1−confidence level). For example, if a 95% CI for the difference in average test scores between two groups is [2, 5], since it doesn't include zero, we can infer that there is a statistically significant difference between the groups at the 0.05 significance level.

If a confidence interval for a single mean does not contain a hypothesized value (from the null hypothesis), it suggests evidence against the null hypothesis at the corresponding alpha level.   



**Evaluating Practical Significance:** Confidence intervals can help in assessing the practical significance or the magnitude of an effect.

Even if a result is statistically significant (small P-value), the confidence interval can show if the effect size is meaningful in the real world. A statistically significant but very narrow confidence interval around a small effect might not be practically important.

Conversely, a wider interval

might still contain a range of values that are practically significant.   


**Comparing Groups or Conditions:** When comparing two groups, overlapping confidence intervals suggest that the population means might not be significantly different. However, non-overlapping intervals indicate a statistically significant difference (though there are nuances, and formal tests are still recommended).


**Important Interpretation Points:**

A confidence level of 95% means that if we were to take many samples and calculate 95% confidence intervals for each, approximately 95% of these intervals would contain the true population parameter.

A specific 95% confidence interval calculated from one sample either contains the true population parameter or it does not. We cannot say there is a 95% probability that the true parameter lies within a specific calculated interval. The probability refers to the process of constructing the interval.

#16 How is Bayes' Theorem used in statistics, and what is its significance?

It provides a mathematical way to update the probability of a hypothesis as new evidence becomes available.   

**The formula for Bayes' Theorem is:**

P(A|B) = (P(B|A) * P(A)) / P(B).

**Where:**

**P(A|B):** The probability of event A happening, given that event B has already happened (conditional probability).

**P(B|A):** The probability of event B happening, given that event A has already happened (likelihood).

**P(A):** The probability of event A happening (prior probability).

**P(B):** The probability of event B happening (evidence).

**How Bayes' Theorem is used in Statistics:**


Bayesian Inference: Bayes' Theorem is the cornerstone of Bayesian statistics. In this approach, probabilities are interpreted as degrees of belief about an event or a parameter. Bayesian inference uses the theorem to update these beliefs in light of new data. It starts with a prior belief about a parameter (represented by a prior probability distribution), incorporates the evidence from the data (through the likelihood function), and produces an updated belief (the posterior probability distribution).   


 **Updating Probabilities:** The core function of Bayes' Theorem is to revise existing probabilities based on new information. This is crucial in many real-world scenarios where our initial understanding needs to be adjusted as we gather more data.   


 **Classification Problems:** In machine learning and statistics, Bayes' Theorem forms the basis of Naive Bayes classifiers, which are used for tasks like spam filtering, text classification, and medical diagnosis. These classifiers calculate the probability of an instance belonging to a particular class given its features, using Bayes' Theorem.   


 **Medical Diagnosis:** Bayes' Theorem can be used to assess the probability of a disease given the results of a diagnostic test. It takes into account the accuracy of the test (sensitivity and specificity) and the prevalence of the disease in the population (prior probability) to give a more accurate probability of having the disease (posterior probability).   


**Risk Assessment:** In finance and insurance, Bayes' Theorem can be used to update the assessment of risks, such as the probability of loan default or the likelihood of an investment being successful, based on new economic indicators or applicant information.   


 **A/B Testing:** While not directly the core calculation, the principles of Bayesian updating are increasingly used in the analysis of A/B tests to provide a more nuanced understanding of the performance of different variations over time.   



**Significance of Bayes' Theorem:**

Incorporates Prior Knowledge: Bayes' Theorem allows for the formal incorporation of prior beliefs or existing knowledge into the statistical analysis. This is particularly useful when data is scarce or when there is strong prior information available.   


**Provides a Framework for Learning:** It offers a natural way to learn from data by sequentially updating probabilities as new evidence arrives. This iterative nature is fundamental to many adaptive systems and learning algorithms.   


**Intuitive Interpretation:** The posterior probability P(A∣B) is often the quantity of direct interest, representing our updated belief about A after observing B. Bayes'

Theorem provides a clear way to calculate this from more readily available probabilities.


**Handles Uncertainty:** By working with probabilities as degrees of belief, Bayes' Theorem provides a powerful framework for reasoning under uncertainty, which is inherent in many real-world problems.


 **Foundation for Bayesian Statistics:** It is the central pillar of Bayesian statistics, which offers a different paradigm for statistical inference compared to classical (frequentist) statistics.

 Bayesian methods are increasingly used in various fields due to their flexibility and interpretability.

#18 What is the Chi-square goodness of fit test, and how is it applied?

   Q18-Answer- The Chi-square goodness of fit test is a statistical hypothesis test used to determine if the observed frequencies of a categorical variable (or a discrete quantitative variable with a finite number of values) match an expected or theoretical distribution. It assesses how well the observed data "fits" a hypothesized distribution.   

**In simpler terms:** Imagine you have a bag of candies with different colors. You have a theory about the proportion of each color in the bag (e.g., 20% red, 30% blue, 50% green). You take a sample of candies and count how many of each color you actually get. The chi-square goodness of fit test helps you determine if your observed counts are significantly different from the counts you would expect based on your theoretical proportions.

**How it is Applied:**

**The chi-square goodness of fit test involves the following steps:**


**State the Hypotheses:**

**Null Hypothesis (H0​):** The observed frequencies are consistent with the hypothesized distribution. In our candy example, this would be that the proportion of colors in the bag matches our 20% red, 30% blue, and 50% green theory.

**Alternative Hypothesis (Ha​):** The observed frequencies are not consistent with the hypothesized distribution. In the candy example, this would be that the proportion of colors in the bag is significantly different from our theoretical proportions.   



 **Determine the Expected Frequencies:** Based on the hypothesized distribution and the total number of observations in your sample, calculate the expected frequency for each category.   


**Formula for Expected Frequency (E):**

E = (Row Total * Column Total) / Grand Total.

Where:

E: Represents the expected frequency for a specific cell in the contingency table.

Row Total: The sum of all values in the row where the cell is located.

Column Total: The sum of all values in the column where the cell is located.

Grand Total: The total number of observations in the entire contingency table.

In our candy example, if we sampled 100 candies, the expected counts would be:


Red: 100×0.20=20
Blue: 100×0.30=30
Green: 100×0.50=50




**Calculate the Chi-square Test Statistic:** This statistic measures the discrepancy between the observed frequencies and the expected frequencies.   

**Formula for Chi-square Statistic (χ2):**

χ² = Σ[(O - E)² / E]

**Where:**

χ2 is the chi-square test statistic.   

Oi​ is the observed frequency for category i.   

Ei​ is the expected frequency for category i.   

k is the number of categories.

**Find the P-value:** Using the calculated chi-square statistic and the degrees of freedom, you can find the p-value from a chi-square distribution table or using statistical software. The p-value represents the probability of observing data as extreme as, or more extreme than, the observed data if the null hypothesis were true.   



**Make a Decision:** Compare the p-value to a chosen significance level (α), typically 0.05.

**If p-value ≤α:** Reject the null hypothesis. There is statistically significant evidence that the observed frequencies do not fit the hypothesized distribution.

**If p-value >α:** Fail to reject the null hypothesis. There is not enough statistically significant evidence to conclude that the observed frequencies are different from the hypothesized distribution.


**Assumptions of the Chi-square Goodness of Fit Test:**

**Categorical Data:** The data must be counts of observations in distinct categories.   

**Independence of Observations:** Each observation should be independent of the others.

**Expected Frequencies:** Generally, all expected frequencies should be at least 1, and the majority (at least 80%) should be 5 or greater. If this assumption is violated, the chi-square approximation may not be accurate.

#20 What is an ANOVA test, and what are its assumptions?

   Q20-Answer- The Analysis of Variance (ANOVA) test is a statistical method used to compare the means of two or more groups to determine if there are any statistically significant differences between them. Unlike t-tests, which are limited to comparing two groups, ANOVA can handle multiple groups simultaneously.   

The core idea behind ANOVA is to examine the variance within each group compared to the variance between the groups. If the variance between the group means is significantly larger than the variance within the groups, it suggests that the group means are likely different.   

In essence, ANOVA tests the null hypothesis that the means of all groups are equal against the alternative hypothesis that at least one group mean is different from the others.   

**There are different types of ANOVA depending on the number of independent variables (factors) being examined:**   

**One-way ANOVA:** Used when there is only one independent variable with two or more levels (groups). It examines the effect of this single factor on the dependent variable.   

**Two-way ANOVA:** Used when there are two independent variables, and it examines the main effect of each independent variable on the dependent variable, as well as the interaction effect between the two independent variables.   

**N-way ANOVA:** An extension for more than two independent variables.

**How ANOVA Works (Simplified):**

 **Calculate the overall mean:** The mean of all the data points across all groups.   

 **Calculate the variance within groups:** A measure of the spread or variability of data points within each individual group.   

 **Calculate the variance between groups:** A measure of the spread or variability of the means of the different groups around the overall mean.   

**Compute the F-statistic:** This is the ratio of the variance between groups to the variance within groups. A large F-statistic suggests that the differences between group means are likely significant.   

**Determine the p-value:** The probability of obtaining an F-statistic as extreme as, or more extreme than, the one calculated, assuming the null hypothesis is true.

 **Make a decision:** If the p-value is less than a chosen significance level (alpha, typically 0.05), we reject the null hypothesis and conclude that there is a statistically significant difference between at least one of the group means.   

**Assumptions of ANOVA:**

**For the results of an ANOVA test to be valid and reliable, several assumptions should be met:**

 **Independence of Observations:** The data points within each group and between groups should be independent of each other. This means that one observation should not influence another. This is often related to the study design and how data is collected.   


**Normality:** The dependent variable should be approximately normally distributed for each level (group) of the independent variable. While ANOVA is somewhat robust to minor violations of normality, especially with larger sample sizes, significant deviations can affect the test's accuracy. This can be assessed using statistical tests (e.g., Shapiro-Wilk, Kolmogorov-Smirnov), histograms, or Q-Q plots of the residuals (the differences between the observed and predicted values).   


 **Homogeneity of Variances (Homoscedasticity):** The variances of the dependent variable should be approximately equal across all groups. Unequal variances (heteroscedasticity) can lead to an increased risk of Type I or Type II errors. This assumption can be tested using statistical tests like Levene's test or Bartlett's test. If this assumption is violated, adjustments to the ANOVA or alternative tests (like Welch's ANOVA) might be necessary.

The interaction effect between the two independent variables on the dependent variable (i.e., does the effect of one factor depend on the level of the other factor?).

**Example:** Investigating the impact of age group, income level, and education level on consumer spending.

**Example:** Measuring the blood pressure of the same group of patients after taking three different dosages of a medication.

**Other Variations:**

**Analysis of Covariance (ANCOVA):** This is a variation of ANOVA that includes one or more covariates (continuous variables) that might influence the dependent variable.

ANCOVA statistically controls for the effects of these covariates to get a clearer picture of the relationship between the independent variable(s) and the dependent variable.   


**Multivariate Analysis of Variance (MANOVA):** This is used when you have two or more dependent variables that are being examined simultaneously across different groups defined by one or more independent variables.   


**Welch's ANOVA:** This is a variation of the one-way ANOVA that does not assume homogeneity of variances between the groups. It's used when the assumption of equal variances is violated.

 **Determining the Degrees of Freedom:** The F-distribution has two types of degrees of freedom: the degrees of freedom for the numerator (df1​) and the degrees of freedom for the denominator (df2​). These depend on the sample sizes and the structure of the data in the specific test.   


**Finding the P-value or Critical Value:**

**P-value approach:** Once the F-statistic is calculated, you compare it to the F-distribution with the appropriate degrees of freedom to find the p-value. The p-value is the probability of observing an F-statistic as extreme as, or more extreme than, the one calculated, assuming the null hypothesis is true.   

**Critical value approach:** You can also choose a significance level (α) and find the critical F-value from an F-distribution table or statistical software based on α and the degrees of freedom.   



**Making a Decision:**

**P-value approach:** If the p-value is less than or equal to the chosen significance level (α), you reject the null hypothesis. This suggests that the observed data provides enough evidence to conclude that the variances are different (in a two-variance test) or that there is a significant difference between the group means (in ANOVA) or that the regression model is significant (in regression).   

**Critical value approach:** If the calculated F-statistic is greater than the critical F-value, you reject the null hypothesis.

```python
#1 Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

import numpy as np
from scipy import stats

def perform_z_test(sample_mean, sample_std, sample_size, population_mean, alpha=0.05, alternative='two-sided'):
    """
    Performs a one-sample Z-test to compare a sample mean to a known population mean.

    Args:
        sample_mean (float): The mean of the sample.
        sample_std (float): The standard deviation of the sample.
        sample_size (int): The size of the sample.
        population_mean (float): The known mean of the population.
        alpha (float, optional): The significance level. Defaults to 0.05.
        alternative (str, optional): The type of alternative hypothesis.
                                     Can be 'two-sided', 'less', or 'greater'.
                                     Defaults to 'two-sided'.

    Returns:
        tuple: A tuple containing the z-statistic and the p-value.
    """
    # Calculate the standard error of the mean
    standard_error = sample_std / np.sqrt(sample_size)

    # Calculate the Z-statistic
    z_statistic = (sample_mean - population_mean) / standard_error

    # Calculate the p-value based on the alternative hypothesis
    if alternative == 'two-sided':
        p_value = 2 * (1 - stats.norm.cdf(np.abs(z_statistic)))
    elif alternative == 'less':
        p_value = stats.norm.cdf(z_statistic)
    elif alternative == 'greater':
        p_value = 1 - stats.norm.cdf(z_statistic)
    else:
        raise ValueError("Invalid alternative hypothesis. Choose 'two-sided', 'less', or 'greater'.")

    return z_statistic, p_value

def interpret_z_test(z_statistic, p_value, alpha, alternative, population_mean, sample_mean):
    """
    Interprets the results of the Z-test.

    Args:
        z_statistic (float): The calculated Z-statistic.
        p_value (float): The calculated p-value.
        alpha (float): The significance level.
        alternative (str): The type of alternative hypothesis.
        population_mean (float): The known population mean.
        sample_mean (float): The mean of the sample.
    """
    print("\n--- Z-Test Results ---")
    print(f"Z-Statistic: {z_statistic:.3f}")
    print(f"P-Value: {p_value:.3f}")
    print(f"Significance Level (alpha): {alpha}")
    print(f"Alternative Hypothesis: {alternative}")
    print(f"Population Mean (μ₀): {population_mean}")
    print(f"Sample Mean (x̄): {sample_mean}")

    if p_value < alpha:
        print("\nConclusion:")
        print(f"The p-value ({p_value:.3f}) is less than the significance level ({alpha}).")
        print("We reject the null hypothesis.")
        if alternative == 'two-sided':
            print(f"There is statistically significant evidence that the sample mean is different from the population mean.")
        elif alternative == 'less':
            print(f"There is statistically significant evidence that the sample mean is less than the population mean.")
        elif alternative == 'greater':
            print(f"There is statistically significant evidence that the sample mean is greater than the population mean.")
    else:
        print("\nConclusion:")
        print(f"The p-value ({p_value:.3f}) is greater than or equal to the significance level ({alpha}).")
        print("We fail to reject the null hypothesis.")
        if alternative == 'two-sided':
            print(f"There is not enough statistically significant evidence to conclude that the sample mean is different from the population mean.")
        elif alternative == 'less':
            print(f"There is not enough statistically significant evidence to conclude that the sample mean is less than the population mean.")
        elif alternative == 'greater':
            print(f"There is not enough statistically significant evidence to conclude that the sample mean is greater than the population mean.")

if __name__ == "__main__":
    # Example usage:
    population_mean = 50  # Known population mean
    sample_data = np.array([52, 55, 48, 49, 51, 53, 50, 54, 47, 56])
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)  # Use sample standard deviation (Bessel's correction)
    sample_size = len(sample_data)
    alpha = 0.05
    alternative_hypothesis = 'two-sided'  # You can change this to 'less' or 'greater'

    # Perform the Z-test
    z_statistic, p_value = perform_z_test(sample_mean, sample_std, sample_size, population_mean, alpha, alternative_hypothesis)

    # Interpret the results
    interpret_z_test(z_statistic, p_value, alpha, alternative_hypothesis, population_mean, sample_mean)

    print("\nNote:")
    print("This Z-test assumes that either the population standard deviation is known,")
    print("or the sample size is large enough (typically n > 30) for the sample standard deviation")
    print("to be a good estimate of the population standard deviation.")
    print("If the population standard deviation is unknown and the sample size is small,")
    print("a t-test would be more appropriate.")
```

```python
#2 Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python.

import numpy as np
from scipy import stats

def perform_hypothesis_test(sample1, sample2, test_type='t_test_ind', alpha=0.05, alternative='two-sided'):
    """
    Performs a hypothesis test on two samples and calculates the p-value.

    Args:
        sample1 (numpy.ndarray): The data for the first sample.
        sample2 (numpy.ndarray): The data for the second sample.
        test_type (str, optional): The type of hypothesis test to perform.
                                   Options: 't_test_ind' (independent samples t-test),
                                            't_test_rel' (related samples t-test),
                                            'z_test_ind' (independent samples z-test - requires known pop std),
                                            'mannwhitneyu' (Mann-Whitney U test - non-parametric).
                                   Defaults to 't_test_ind'.
        alpha (float, optional): The significance level. Defaults to 0.05.
        alternative (str, optional): The type of alternative hypothesis.
                                     Can be 'two-sided', 'less', or 'greater'.
                                     Defaults to 'two-sided'.

    Returns:
        tuple: A tuple containing the test statistic and the p-value.
               Returns None, None if an invalid test_type is provided.
    """
    if test_type == 't_test_ind':
        statistic, p_value = stats.ttest_ind(sample1, sample2, alternative=alternative)
    elif test_type == 't_test_rel':
        statistic, p_value = stats.ttest_rel(sample1, sample2, alternative=alternative)
    elif test_type == 'z_test_ind':
        # For Z-test, we need to assume known population standard deviations.
        # Here, we'll use the sample standard deviations as an approximation (for demonstration).
        # In a real scenario, you should have the population std.
        n1 = len(sample1)
        n2 = len(sample2)
        mean1 = np.mean(sample1)
        mean2 = np.mean(sample2)
        std_err_diff = np.sqrt((np.std(sample1, ddof=1)**2 / n1) + (np.std(sample2, ddof=1)**2 / n2))
        statistic = (mean1 - mean2) / std_err_diff
        if alternative == 'two-sided':
            p_value = 2 * (1 - stats.norm.cdf(np.abs(statistic)))
        elif alternative == 'less':
            p_value = stats.norm.cdf(statistic)
        elif alternative == 'greater':
            p_value = 1 - stats.norm.cdf(statistic)
        else:
            raise ValueError("Invalid alternative hypothesis for z_test_ind.")
    elif test_type == 'mannwhitneyu':
        statistic, p_value = stats.mannwhitneyu(sample1, sample2, alternative=alternative)
    else:
        print(f"Error: Invalid test_type '{test_type}'. Choose from 't_test_ind', 't_test_rel', 'z_test_ind', or 'mannwhitneyu'.")
        return None, None

    return statistic, p_value

def interpret_p_value(p_value, alpha):
    """
    Interprets the p-value based on the significance level.

    Args:
        p_value (float): The calculated p-value.
        alpha (float): The significance level.

    Returns:
        str: The interpretation of the p-value.
    """
    if p_value is not None:
        if p_value < alpha:
            return f"The p-value ({p_value:.3f}) is less than the significance level ({alpha}). We reject the null hypothesis."
        else:
            return f"The p-value ({p_value:.3f}) is greater than or equal to the significance level ({alpha}). We fail to reject the null hypothesis."
    else:
        return "No p-value to interpret due to an error in the test."

if __name__ == "__main__":
    # --- Simulate Random Data ---
    np.random.seed(42)  # for reproducibility
    sample_size1 = 30
    sample_size2 = 35
    mean1 = 5
    mean2 = 5.5
    std_dev1 = 1
    std_dev2 = 1.2

    sample_group1 = np.random.normal(mean1, std_dev1, sample_size1)
    sample_group2 = np.random.normal(mean2, std_dev2, sample_size2)

    print("Sample Group 1 (first 10 elements):", sample_group1[:10])
    print("Sample Group 2 (first 10 elements):", sample_group2[:10])

    # --- Choose Hypothesis Test Type ---
    test_type_to_use = 't_test_ind'  # Try 't_test_rel' if samples are related, 'z_test_ind' with caution, 'mannwhitneyu' for non-parametric
    alpha_level = 0.05
    alternative_hypothesis = 'two-sided' # Try 'less' or 'greater'

    # --- Perform Hypothesis Test ---
    statistic, p_value = perform_hypothesis_test(sample_group1, sample_group2, test_type_to_use, alpha_level, alternative_hypothesis)

    print(f"\n--- Results of {test_type_to_use} ---")
    if statistic is not None:
        print(f"Test Statistic: {statistic:.3f}")
    if p_value is not None:
        print(f"P-value: {p_value:.3f}")

    # --- Interpret the P-value ---
    interpretation = interpret_p_value(p_value, alpha_level)
    print("\nInterpretation:", interpretation)
```

```python
#3 Implement a one-sample Z-test using Python to compare the sample mean with the population mean.

import numpy as np
from scipy import stats

def one_sample_z_test(sample_data, population_mean, population_std, alpha=0.05, alternative='two-sided'):
    """
    Performs a one-sample Z-test to compare the mean of a sample to a known population mean
    when the population standard deviation is known.

    Args:
        sample_data (numpy.ndarray or list): The data of the sample.
        population_mean (float): The known mean of the population.
        population_std (float): The known standard deviation of the population.
        alpha (float, optional): The significance level. Defaults to 0.05.
        alternative (str, optional): The type of alternative hypothesis.
                                     Can be 'two-sided', 'less', or 'greater'.
                                     Defaults to 'two-sided'.

    Returns:
        tuple: A tuple containing the z-statistic and the p-value.
    """
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)
    standard_error = population_std / np.sqrt(sample_size)
    z_statistic = (sample_mean - population_mean) / standard_error

    if alternative == 'two-sided':
        p_value = 2 * (1 - stats.norm.cdf(np.abs(z_statistic)))
    elif alternative == 'less':
        p_value = stats.norm.cdf(z_statistic)
    elif alternative == 'greater':
        p_value = 1 - stats.norm.cdf(z_statistic)
    else:
        raise ValueError("Invalid alternative hypothesis. Choose 'two-sided', 'less', or 'greater'.")

    return z_statistic, p_value

def interpret_z_test_result(z_statistic, p_value, alpha, alternative, population_mean, sample_mean):
    """
    Interprets the results of the one-sample Z-test.

    Args:
        z_statistic (float): The calculated Z-statistic.
        p_value (float): The calculated p-value.
        alpha (float): The significance level.
        alternative (str): The type of alternative hypothesis.
        population_mean (float): The known population mean.
        sample_mean (float): The mean of the sample.
    """
    print("\n--- One-Sample Z-Test Results ---")
    print(f"Z-Statistic: {z_statistic:.3f}")
    print(f"P-Value: {p_value:.3f}")
    print(f"Significance Level (alpha): {alpha}")
    print(f"Alternative Hypothesis: {alternative}")
    print(f"Population Mean (μ₀): {population_mean}")
    print(f"Sample Mean (x̄): {sample_mean:.3f}")

    if p_value < alpha:
        print("\nConclusion:")
        print(f"The p-value ({p_value:.3f}) is less than the significance level ({alpha}).")
        print("We reject the null hypothesis.")
        if alternative == 'two-sided':
            print(f"There is statistically significant evidence that the sample mean is different from the population mean.")
        elif alternative == 'less':
            print(f"There is statistically significant evidence that the sample mean is less than the population mean.")
        elif alternative == 'greater':
            print(f"There is statistically significant evidence that the sample mean is greater than the population mean.")
    else:
        print("\nConclusion:")
        print(f"The p-value ({p_value:.3f}) is greater than or equal to the significance level ({alpha}).")
        print("We fail to reject the null hypothesis.")
        if alternative == 'two-sided':
            print(f"There is not enough statistically significant evidence to conclude that the sample mean is different from the population mean.")
        elif alternative == 'less':
            print(f"There is not enough statistically significant evidence to conclude that the sample mean is less than the population mean.")
        elif alternative == 'greater':
            print(f"There is not enough statistically significant evidence to conclude that the sample mean is greater than the population mean.")

if __name__ == "__main__":
    # Example Usage:
    population_mean = 100  # Known population mean
    population_std = 15   # Known population standard deviation
    sample_data = np.array([105, 110, 98, 102, 108, 95, 103, 112, 99, 106])
    alpha = 0.05
    alternative_hypothesis = 'two-sided'  # You can change this to 'less' or 'greater'

    # Perform the one-sample Z-test
    z_statistic, p_value = one_sample_z_test(sample_data, population_mean, population_std, alpha, alternative_hypothesis)

    # Interpret the results
    sample_mean = np.mean(sample_data)
    interpret_z_test_result(z_statistic, p_value, alpha, alternative_hypothesis, population_mean, sample_mean)
```

```python
#4 Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def two_tailed_z_test_and_visualization(sample_mean, sample_size, population_mean, population_std, alpha=0.05):
    """
    Performs a two-tailed Z-test and visualizes the decision region.

    Args:
        sample_mean (float): The mean of the sample.
        sample_size (int): The size of the sample.
        population_mean (float): The known mean of the population.
        population_std (float): The known standard deviation of the population.
        alpha (float, optional): The significance level. Defaults to 0.05.

    Returns:
        tuple: A tuple containing the z-statistic and the p-value.
    """
    # Calculate the standard error of the mean
    standard_error = population_std / np.sqrt(sample_size)

    # Calculate the Z-statistic
    z_statistic = (sample_mean - population_mean) / standard_error

    # Calculate the p-value for a two-tailed test
    p_value = 2 * (1 - stats.norm.cdf(np.abs(z_statistic)))

    # Determine the critical Z-values for a two-tailed test
    critical_z = stats.norm.ppf(1 - alpha / 2)
    critical_z_lower = -critical_z

    # --- Visualization ---
    x = np.linspace(population_mean - 4 * standard_error, population_mean + 4 * standard_error, 200)
    y = stats.norm.pdf(x, population_mean, standard_error)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label=f'Sampling Distribution (μ={population_mean}, SE={standard_error:.2f})')

    # Shade the rejection regions
    plt.fill_between(x, y, where=(x <= critical_z_lower), color='red', alpha=0.3, label=f'Rejection Region (Z <= {critical_z_lower:.2f})')
    plt.fill_between(x, y, where=(x >= critical_z), color='red', alpha=0.3, label=f'Rejection Region (Z >= {critical_z:.2f})')

    # Plot the sample mean
    plt.axvline(sample_mean, color='blue', linestyle='--', linewidth=1.5, label=f'Sample Mean = {sample_mean:.2f}')

    # Plot the critical Z-values
    plt.axvline(population_mean + critical_z_lower * standard_error, color='gray', linestyle='--', linewidth=1, label=f'Critical Value Lower = {population_mean + critical_z_lower * standard_error:.2f} (Z={critical_z_lower:.2f})')
    plt.axvline(population_mean + critical_z * standard_error, color='gray', linestyle='--', linewidth=1, label=f'Critical Value Upper = {population_mean + critical_z * standard_error:.2f} (Z={critical_z:.2f})')

    plt.title('Two-Tailed Z-Test Visualization')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    return z_statistic, p_value

def interpret_z_test_result(z_statistic, p_value, alpha):
    """
    Interprets the results of the two-tailed Z-test.

    Args:
        z_statistic (float): The calculated Z-statistic.
        p_value (float): The calculated p-value.
        alpha (float): The significance level.
    """
    print("\n--- Two-Tailed Z-Test Results ---")
    print(f"Z-Statistic: {z_statistic:.3f}")
    print(f"P-Value: {p_value:.3f}")
    print(f"Significance Level (alpha): {alpha}")

    if p_value < alpha:
        print("\nConclusion:")
        print(f"The p-value ({p_value:.3f}) is less than the significance level ({alpha}).")
        print("We reject the null hypothesis.")
        print("There is statistically significant evidence that the sample mean is different from the population mean.")
    else:
        print("\nConclusion:")
        print(f"The p-value ({p_value:.3f}) is greater than or equal to the significance level ({alpha}).")
        print("We fail to reject the null hypothesis.")
        print("There is not enough statistically significant evidence to conclude that the sample mean is different from the population mean.")

if __name__ == "__main__":
    # Example Usage:
    population_mean = 50
    population_std = 10
    sample_mean = 53
    sample_size = 40
    alpha = 0.05

    # Perform the two-tailed Z-test and visualize
    z_statistic, p_value = two_tailed_z_test_and_visualization(sample_mean, sample_size, population_mean, population_std, alpha)

    # Interpret the results
    interpret_z_test_result(z_statistic, p_value, alpha)
```

```python
#5 Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def visualize_hypothesis_errors(null_mean, alt_mean, std_dev, sample_size, alpha=0.05, power=0.8):
    """
    Calculates and visualizes Type 1 and Type 2 errors in hypothesis testing
    for a one-sample Z-test (for simplicity of visualization).

    Args:
        null_mean (float): The mean under the null hypothesis (H0).
        alt_mean (float): The mean under the alternative hypothesis (Ha).
        std_dev (float): The population standard deviation (assumed known for Z-test).
        sample_size (int): The size of the sample.
        alpha (float, optional): The significance level (probability of Type 1 error). Defaults to 0.05.
        power (float, optional): The desired power of the test (1 - probability of Type 2 error). Defaults to 0.8.
    """
    if null_mean == alt_mean:
        raise ValueError("Null and alternative means cannot be the same for visualization.")

    standard_error = std_dev / np.sqrt(sample_size)

    # --- Calculate Critical Value (for a two-tailed test) ---
    critical_z = stats.norm.ppf(1 - alpha / 2)
    critical_value_upper_H0 = null_mean + critical_z * standard_error
    critical_value_lower_H0 = null_mean - critical_z * standard_error

    # --- Calculate Beta (Probability of Type 2 Error) and Power (1 - Beta) ---
    # We need to find the critical region under H0 and then calculate the probability
    # of the alternative distribution falling within the "fail to reject" region of H0.

    # For simplicity, let's calculate Beta based on the desired power.
    # In a real scenario, you might iterate on sample size to achieve desired power.
    beta = 1 - power

    # --- Visualization ---
    x = np.linspace(min(null_mean, alt_mean) - 4 * standard_error,
                    max(null_mean, alt_mean) + 4 * standard_error, 200)
    y_null = stats.norm.pdf(x, null_mean, standard_error)
    y_alt = stats.norm.pdf(x, alt_mean, standard_error)

    plt.figure(figsize=(12, 7))

    # Plot the null distribution
    plt.plot(x, y_null, color='blue', label=f'Null Hypothesis (μ={null_mean})')
    plt.fill_between(x, y_null, where=((x >= critical_value_lower_H0) & (x <= critical_value_upper_H0)),
                     color='lightgray', alpha=0.7, label='Fail to Reject H0')
    plt.fill_between(x, y_null, where=(x < critical_value_lower_H0), color='red', alpha=0.3, label='Reject H0 (Type 1 Error Area)')
    plt.fill_between(x, y_null, where=(x > critical_value_upper_H0), color='red', alpha=0.3)

    # Plot the alternative distribution
    plt.plot(x, y_alt, color='green', label=f'Alternative Hypothesis (μ={alt_mean})')
    plt.fill_between(x, y_alt, where=((x >= critical_value_lower_H0) & (x <= critical_value_upper_H0)),
                     color='yellow', alpha=0.5, label='Type 2 Error Area (Fail to Reject H0 when Ha is true)')

    # Add vertical lines for critical values
    plt.axvline(critical_value_lower_H0, color='red', linestyle='--', linewidth=1, label=f'Critical Value (Lower) = {critical_value_lower_H0:.2f}')
    plt.axvline(critical_value_upper_H0, color='red', linestyle='--', linewidth=1, label=f'Critical Value (Upper) = {critical_value_upper_H0:.2f}')

    plt.title('Visualization of Type 1 and Type 2 Errors')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.annotate(f'Alpha (Type 1 Error Rate) = {alpha}', xy=(min(x), max(y_null) * 0.9), color='red')
    plt.annotate(f'Beta (Type 2 Error Rate) ≈ {beta:.3f}', xy=(min(x), max(y_null) * 0.8), color='orange')
    plt.annotate(f'Power (1 - Beta) ≈ {power:.3f}', xy=(min(x), max(y_null) * 0.7), color='green')
    plt.show()

    print("\n--- Type 1 and Type 2 Error Analysis ---")
    print(f"Null Hypothesis Mean (μ₀): {null_mean}")
    print(f"Alternative Hypothesis Mean (μa): {alt_mean}")
    print(f"Population Standard Deviation (σ): {std_dev}")
    print(f"Sample Size (n): {sample_size}")
    print(f"Significance Level (Alpha): {alpha}")
    print(f"Critical Values under H0: [{critical_value_lower_H0:.2f}, {critical_value_upper_H0:.2f}]")
    print(f"Probability of Type 1 Error (Alpha): {alpha}")
    print(f"Approximate Probability of Type 2 Error (Beta): {beta:.3f}")
    print(f"Approximate Power (1 - Beta): {power:.3f}")

if __name__ == "__main__":
    # Example Usage:
    null_mean = 50
    alt_mean = 53
    std_dev = 10
    sample_size = 30
    alpha = 0.05
    power = 0.7  # Desired power (adjust to see the effect on Beta)

    visualize_hypothesis_errors(null_mean, alt_mean, std_dev, sample_size, alpha, power)
```

```python
#6 Write a Python program to perform an independent T-test and interpret the results.

import numpy as np
from scipy import stats

def independent_t_test(sample1, sample2, alpha=0.05, equal_var=True, alternative='two-sided'):
    """
    Performs an independent samples t-test (Student's t-test) to compare the means of two
    independent groups.

    Args:
        sample1 (numpy.ndarray or list): The data for the first sample.
        sample2 (numpy.ndarray or list): The data for the second sample.
        alpha (float, optional): The significance level. Defaults to 0.05.
        equal_var (bool, optional): A boolean indicating whether to assume equal population variances.
                                     If True (default), a pooled variance t-test is performed.
                                     If False, Welch's t-test (for unequal variances) is performed.
        alternative (str, optional): The type of alternative hypothesis.
                                     Can be 'two-sided', 'less', or 'greater'.
                                     Defaults to 'two-sided'.

    Returns:
        tuple: A tuple containing the t-statistic and the p-value.
    """
    statistic, p_value = stats.ttest_ind(sample1, sample2, equal_var=equal_var, alternative=alternative)
    return statistic, p_value

def interpret_independent_t_test(statistic, p_value, alpha, equal_var, alternative, mean1, mean2):
    """
    Interprets the results of the independent samples t-test.

    Args:
        statistic (float): The calculated t-statistic.
        p_value (float): The calculated p-value.
        alpha (float): The significance level.
        equal_var (bool): Indicates whether equal variances were assumed.
        alternative (str): The type of alternative hypothesis.
        mean1 (float): The mean of the first sample.
        mean2 (float): The mean of the second sample.
    """
    print("\n--- Independent Samples T-Test Results ---")
    print(f"T-Statistic: {statistic:.3f}")
    print(f"P-Value: {p_value:.3f}")
    print(f"Significance Level (alpha): {alpha}")
    print(f"Assumed Equal Variances: {equal_var}")
    print(f"Alternative Hypothesis: {alternative}")
    print(f"Sample 1 Mean (x̄₁): {mean1:.3f}")
    print(f"Sample 2 Mean (x̄₂): {mean2:.3f}")

    if p_value < alpha:
        print("\nConclusion:")
        print(f"The p-value ({p_value:.3f}) is less than the significance level ({alpha}).")
        print("We reject the null hypothesis.")
        if alternative == 'two-sided':
            print(f"There is statistically significant evidence that the means of the two groups are different.")
        elif alternative == 'less':
            print(f"There is statistically significant evidence that the mean of the first group is less than the mean of the second group.")
        elif alternative == 'greater':
            print(f"There is statistically significant evidence that the mean of the first group is greater than the mean of the second group.")
    else:
        print("\nConclusion:")
        print(f"The p-value ({p_value:.3f}) is greater than or equal to the significance level ({alpha}).")
        print("We fail to reject the null hypothesis.")
        if alternative == 'two-sided':
            print(f"There is not enough statistically significant evidence to conclude that the means of the two groups are different.")
        elif alternative == 'less':
            print(f"There is not enough statistically significant evidence to conclude that the mean of the first group is not significantly less than the mean of the second group.")
        elif alternative == 'greater':
            print(f"There is not enough statistically significant evidence to conclude that the mean of the first group is not significantly greater than the mean of the second group.")

if __name__ == "__main__":
    # Example Usage:
    np.random.seed(0)  # for reproducibility
    group1 = np.random.normal(loc=5, scale=2, size=50)
    group2 = np.random.normal(loc=5.5, scale=2.5, size=60)
    alpha = 0.05
    assume_equal_variances = True  # Set to False if you suspect unequal variances
    alternative_hypothesis = 'two-sided'  # You can change this to 'less' or 'greater'

    # Perform the independent t-test
    t_statistic, p_value = independent_t_test(group1, group2, alpha, assume_equal_variances, alternative_hypothesis)

    # Interpret the results
    mean1 = np.mean(group1)
    mean2 = np.mean(group2)
    interpret_independent_t_test(t_statistic, p_value, alpha, assume_equal_variances, alternative_hypothesis, mean1, mean2)

    print("\nNote:")
    print("The independent samples t-test assumes that the data in each group is approximately normally distributed.")
    print("If the normality assumption is severely violated, consider using a non-parametric test like the Mann-Whitney U test.")
    print("The choice between assuming equal variances (Student's t-test) and not assuming equal variances (Welch's t-test)")
    print("can be informed by conducting a test for equality of variances (e.g., Levene's test).")
```

```python
#7 Perform a paired sample T-test using Python and visualize the comparison results.

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def paired_sample_t_test_and_visualization(before, after, alpha=0.05, alternative='two-sided'):
    """
    Performs a paired (related) samples t-test and visualizes the comparison results.

    Args:
        before (numpy.ndarray or list): Data for the first measurement (e.g., before treatment).
        after (numpy.ndarray or list): Data for the second measurement (e.g., after treatment),
                                       corresponding to the 'before' data.
        alpha (float, optional): The significance level. Defaults to 0.05.
        alternative (str, optional): The type of alternative hypothesis.
                                     Can be 'two-sided', 'less', or 'greater'.
                                     Defaults to 'two-sided'.

    Returns:
        tuple: A tuple containing the t-statistic and the p-value.
    """
    if len(before) != len(after):
        raise ValueError("The 'before' and 'after' datasets must have the same number of observations for a paired t-test.")

    # Calculate the difference between the paired observations
    differences = np.array(after) - np.array(before)
    mean_difference = np.mean(differences)
    std_difference = np.std(differences, ddof=1)  # Sample standard deviation
    n = len(differences)
    standard_error_difference = std_difference / np.sqrt(n)

    # Calculate the t-statistic
    t_statistic = mean_difference / standard_error_difference

    # Calculate the p-value
    df = n - 1  # Degrees of freedom for paired t-test
    if alternative == 'two-sided':
        p_value = 2 * (1 - stats.t.cdf(np.abs(t_statistic), df))
    elif alternative == 'less':
        p_value = stats.t.cdf(t_statistic, df)
    elif alternative == 'greater':
        p_value = 1 - stats.t.cdf(t_statistic, df)
    else:
        raise ValueError("Invalid alternative hypothesis. Choose 'two-sided', 'less', or 'greater'.")

    # --- Visualization ---
    plt.figure(figsize=(10, 6))

    # Scatter plot of before vs. after
    plt.scatter(before, after, alpha=0.7, label='Individual Pairs')
    plt.xlabel('Before Measurement')
    plt.ylabel('After Measurement')
    plt.title('Comparison of Paired Measurements')

    # Plot the line y = x for reference (no change)
    min_val = min(min(before), min(after))
    max_val = max(max(before), max(after))
    plt.plot([min_val, max_val], [min_val, max_val], color='gray', linestyle='--', linewidth=1, label='No Change')

    # Plot the mean difference
    plt.axvline(np.mean(before), color='blue', linestyle='--', linewidth=1, label=f'Mean Before = {np.mean(before):.2f}')
    plt.axhline(np.mean(after), color='green', linestyle='--', linewidth=1, label=f'Mean After = {np.mean(after):.2f}')

    plt.legend()
    plt.grid(True)
    plt.show()

    # --- Visualization of Differences ---
    plt.figure(figsize=(10, 6))
    plt.hist(differences, bins='auto', color='skyblue', edgecolor='black', alpha=0.7, label='Differences (After - Before)')
    plt.axvline(mean_difference, color='red', linestyle='-', linewidth=2, label=f'Mean Difference = {mean_difference:.2f}')
    plt.xlabel('Difference (After - Before)')
    plt.ylabel('Frequency')
    plt.title('Distribution of Differences')
    plt.legend()
    plt.grid(True)
    plt.show()

    return t_statistic, p_value

def interpret_paired_t_test(t_statistic, p_value, alpha, alternative, mean_difference):
    """
    Interprets the results of the paired samples t-test.

    Args:
        t_statistic (float): The calculated t-statistic.
        p_value (float): The calculated p-value.
        alpha (float): The significance level.
        alternative (str): The type of alternative hypothesis.
        mean_difference (float): The mean of the differences (After - Before).
    """
    print("\n--- Paired Samples T-Test Results ---")
    print(f"T-Statistic: {t_statistic:.3f}")
    print(f"P-Value: {p_value:.3f}")
    print(f"Significance Level (alpha): {alpha}")
    print(f"Alternative Hypothesis: {alternative}")
    print(f"Mean Difference (After - Before): {mean_difference:.3f}")

    if p_value < alpha:
        print("\nConclusion:")
        print(f"The p-value ({p_value:.3f}) is less than the significance level ({alpha}).")
        print("We reject the null hypothesis.")
        if alternative == 'two-sided':
            print(f"There is statistically significant evidence that there is a difference between the 'before' and 'after' measurements.")
        elif alternative == 'less':
            print(f"There is statistically significant evidence that the 'after' measurements are significantly less than the 'before' measurements.")
        elif alternative == 'greater':
            print(f"There is statistically significant evidence that the 'after' measurements are significantly greater than the 'before' measurements.")
    else:
        print("\nConclusion:")
        print(f"The p-value ({p_value:.3f}) is greater than or equal to the significance level ({alpha}).")
        print("We fail to reject the null hypothesis.")
        if alternative == 'two-sided':
            print(f"There is not enough statistically significant evidence to conclude that there is a difference between the 'before' and 'after' measurements.")
        elif alternative == 'less':
            print(f"There is not enough statistically significant evidence to conclude that the 'after' measurements are significantly less than the 'before' measurements.")
        elif alternative == 'greater':
            print(f"There is not enough statistically significant evidence to conclude that the 'after' measurements are significantly greater than the 'before' measurements.")

if __name__ == "__main__":
    # Example Usage:
    np.random.seed(0)
    before_treatment = np.random.normal(loc=30, scale=5, size=40)
    after_treatment = before_treatment + np.random.normal(loc=2, scale=3, size=40) # Simulate an effect

    alpha = 0.05
    alternative_hypothesis = 'greater'  # Test if 'after' is greater than 'before'

    # Perform the paired sample t-test and visualize
    t_statistic, p_value = paired_sample_t_test_and_visualization(before_treatment, after_treatment, alpha, alternative_hypothesis)

    # Interpret the results
    mean_difference = np.mean(np.array(after_treatment) - np.array(before_treatment))
    interpret_paired_t_test(t_statistic, p_value, alpha, alternative_hypothesis, mean_difference)

    print("\nNote:")
    print("The paired samples t-test assumes that the differences between the paired observations are approximately normally distributed.")
    print("The visualization includes a scatter plot to show individual pairs and a histogram of the differences.")
```

```python
#8 Simulate data and perform both Z-test and T-test, then compare the results using Python.

import numpy as np
from scipy import stats

def perform_z_test(sample_mean, sample_std, sample_size, population_mean, alpha=0.05, alternative='two-sided'):
    """Performs a one-sample Z-test."""
    standard_error = population_std / np.sqrt(sample_size)
    z_statistic = (sample_mean - population_mean) / standard_error
    if alternative == 'two-sided':
        p_value_z = 2 * (1 - stats.norm.cdf(np.abs(z_statistic)))
    elif alternative == 'less':
        p_value_z = stats.norm.cdf(z_statistic)
    elif alternative == 'greater':
        p_value_z = 1 - stats.norm.cdf(z_statistic)
    return z_statistic, p_value_z

def perform_t_test(sample_data, population_mean, alpha=0.05, alternative='two-sided'):
    """Performs a one-sample T-test."""
    sample_mean_t = np.mean(sample_data)
    sample_std_t = np.std(sample_data, ddof=1)  # Sample standard deviation
    sample_size_t = len(sample_data)
    standard_error_t = sample_std_t / np.sqrt(sample_size_t)
    t_statistic = (sample_mean_t - population_mean) / standard_error_t
    df = sample_size_t - 1
    if alternative == 'two-sided':
        p_value_t = 2 * (1 - stats.t.cdf(np.abs(t_statistic), df))
    elif alternative == 'less':
        p_value_t = stats.t.cdf(t_statistic, df)
    elif alternative == 'greater':
        p_value_t = 1 - stats.t.cdf(t_statistic, df)
    return t_statistic, p_value_t

def compare_z_and_t_results(z_stat, p_val_z, t_stat, p_val_t, alpha):
    """Compares and prints the results of the Z-test and T-test."""
    print("\n--- Comparison of Z-test and T-test Results ---")
    print(f"Significance Level (alpha): {alpha}")
    print(f"\nZ-test:")
    print(f"  Z-Statistic: {z_stat:.4f}")
    print(f"  P-Value: {p_val_z:.4f}")
    if p_val_z < alpha:
        print("  Conclusion: Reject the null hypothesis.")
    else:
        print("  Conclusion: Fail to reject the null hypothesis.")

    print(f"\nT-test:")
    print(f"  T-Statistic: {t_stat:.4f}")
    print(f"  P-Value: {p_val_t:.4f}")
    if p_val_t < alpha:
        print("  Conclusion: Reject the null hypothesis.")
    else:
        print("  Conclusion: Fail to reject the null hypothesis.")

    print("\n--- Observations ---")
    if p_val_z < alpha and p_val_t >= alpha:
        print("The Z-test rejected the null hypothesis, while the T-test failed to reject it.")
        print("This discrepancy might occur with small sample sizes or when the sample standard deviation")
        print("is considerably different from the population standard deviation.")
    elif p_val_z >= alpha and p_val_t < alpha:
        print("The T-test rejected the null hypothesis, while the Z-test failed to reject it.")
        print("This is less common but could happen due to sampling variability, especially with smaller samples.")
    elif p_val_z < alpha and p_val_t < alpha:
        print("Both the Z-test and T-test rejected the null hypothesis.")
    else:
        print("Both the Z-test and T-test failed to reject the null hypothesis.")
    print("The T-test generally accounts for the uncertainty introduced by estimating the population")
    print("standard deviation from the sample, especially with smaller sample sizes. As the sample")
    print("size increases, the results of the Z-test (if the population standard deviation is known)")
    print("and the T-test tend to converge due to the Central Limit Theorem.")

if __name__ == "__main__":
    # --- Simulate Data ---
    np.random.seed(42)
    population_mean = 50
    population_std = 10  # Assume population standard deviation is known for Z-test

    sample_size = 30
    sample_data = np.random.normal(loc=52, scale=8, size=sample_size)
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)

    alpha = 0.05
    alternative_hypothesis = 'two-sided'

    # --- Perform Z-test ---
    z_statistic, p_value_z = perform_z_test(sample_mean, population_std, sample_size, population_mean, alpha, alternative_hypothesis)

    # --- Perform T-test ---
    t_statistic, p_value_t = perform_t_test(sample_data, population_mean, alpha, alternative_hypothesis)

    # --- Compare Results ---
    compare_z_and_t_results(z_statistic, p_value_z, t_statistic, p_value_t, alpha)
```

```python
#9 Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
    data (array-like): Sample data
    confidence (float): Confidence level (default: 0.95 for 95% CI)

    Returns:
    tuple: (mean, margin_of_error, (lower_bound, upper_bound))
    """
    # Convert data to numpy array
    data = np.array(data)

    # Calculate sample statistics
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error of the mean

    # Calculate margin of error
    margin = std_err * stats.t.ppf((1 + confidence) / 2, n-1)

    # Calculate confidence interval bounds
    lower = mean - margin
    upper = mean + margin

    return mean, margin, (lower, upper)

# Example usage:
sample_data = [72, 68, 74, 65, 70, 71, 69, 73, 67, 72]
mean, margin, (lower, upper) = confidence_interval(sample_data)
print(f"Mean: {mean:.2f}")
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
print(f"Margin of Error: ±{margin:.2f}")
```

```python
#10 Write a Python program to calculate the margin of error for a given confidence level using sample data.

import numpy as np
from scipy import stats

def calculate_margin_of_error(data, confidence_level=0.95):
    """
    Calculate the margin of error for a sample mean at a given confidence level.

    Parameters:
    data (array-like): The sample data points
    confidence_level (float): The desired confidence level (e.g., 0.95 for 95%)

    Returns:
    dict: A dictionary containing all relevant statistics including margin of error
    """
    # Convert data to numpy array
    data = np.array(data)
    n = len(data)

    # Handle edge cases
    if n < 2:
        raise ValueError("Sample size must be at least 2 to calculate margin of error")

    if not (0 < confidence_level < 1):
        raise ValueError("Confidence level must be between 0 and 1")

    # Calculate basic statistics
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Use ddof=1 for sample standard deviation
    standard_error = sample_std / np.sqrt(n)

    # Determine critical value based on sample size
    if n >= 30:
        # Use z-score for large samples (normal approximation)
        critical_value = stats.norm.ppf((1 + confidence_level) / 2)
    else:
        # Use t-score for small samples
        critical_value = stats.t.ppf((1 + confidence_level) / 2, df=n-1)

    # Calculate margin of error
    margin_of_error = critical_value * standard_error

    # Calculate confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return {
        'sample_size': n,
        'sample_mean': sample_mean,
        'sample_std_dev': sample_std,
        'standard_error': standard_error,
        'confidence_level': confidence_level,
        'critical_value': critical_value,
        'margin_of_error': margin_of_error,
        'confidence_interval': (lower_bound, upper_bound)
    }

def main():
    # Example usage
    sample_data = [72.5, 68.2, 74.1, 65.7, 70.3, 71.0, 69.4, 73.2, 67.8, 72.9]
    confidence = 0.95  # 95% confidence level

    try:
        results = calculate_margin_of_error(sample_data, confidence)

        print("\nMargin of Error Calculation Results:")
        print("=" * 40)
        print(f"Sample Size: {results['sample_size']}")
        print(f"Sample Mean: {results['sample_mean']:.4f}")
        print(f"Sample Standard Deviation: {results['sample_std_dev']:.4f}")
        print(f"Standard Error: {results['standard_error']:.4f}")
        print(f"Confidence Level: {results['confidence_level'] * 100:.0f}%")
        print(f"Critical Value: {results['critical_value']:.4f}")
        print(f"Margin of Error: ±{results['margin_of_error']:.4f}")
        print(f"Confidence Interval: ({results['confidence_interval'][0]:.4f}, "
              f"{results['confidence_interval'][1]:.4f})")

        # Calculate for different confidence levels
        print("\nMargins of Error at Different Confidence Levels:")
        print("-" * 40)
        for cl in [0.90, 0.95, 0.99]:
            res = calculate_margin_of_error(sample_data, cl)
            print(f"{cl * 100:.0f}% CI: ±{res['margin_of_error']:.4f} "
                  f"(Range: {res['confidence_interval'][0]:.2f} to {res['confidence_interval'][1]:.2f})")

    except ValueError as e:
        print(f"Error: {e}")

if __name__ == "__main__":
    main()
```

```python
#11 Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta, binom

class BayesianInference:
    def __init__(self, prior_alpha=1, prior_beta=1):
        """
        Initialize Bayesian model with Beta prior (default is uniform prior)

        Parameters:
        prior_alpha (float): Alpha parameter of Beta prior
        prior_beta (float): Beta parameter of Beta prior
        """
        self.prior_alpha = prior_alpha
        self.prior_beta = prior_beta
        self.posterior_alpha = prior_alpha
        self.posterior_beta = prior_beta
        self.data = {'successes': 0, 'failures': 0}

    def update(self, successes, failures):
        """
        Update posterior distribution with new data

        Parameters:
        successes (int): Number of successful trials
        failures (int): Number of failed trials
        """
        self.data['successes'] += successes
        self.data['failures'] += failures
        self.posterior_alpha = self.prior_alpha + self.data['successes']
        self.posterior_beta = self.prior_beta + self.data['failures']

    def plot_distributions(self):
        """Plot prior, likelihood, and posterior distributions"""
        theta = np.linspace(0, 1, 1000)

        # Prior distribution
        prior = beta(self.prior_alpha, self.prior_beta).pdf(theta)

        # Likelihood (binomial)
        n = self.data['successes'] + self.data['failures']
        k = self.data['successes']
        likelihood = binom.pmf(k, n, theta)

        # Posterior distribution
        posterior = beta(self.posterior_alpha, self.posterior_beta).pdf(theta)

        plt.figure(figsize=(12, 6))
        plt.plot(theta, prior, label='Prior', color='blue')
        plt.plot(theta, likelihood, label='Likelihood', color='green')
        plt.plot(theta, posterior, label='Posterior', color='red')
        plt.title('Bayesian Inference: Prior, Likelihood, and Posterior')
        plt.xlabel('Parameter θ')
        plt.ylabel('Probability Density')
        plt.legend()
        plt.grid(True)
        plt.show()

    def get_posterior_stats(self):
        """Get posterior statistics"""
        mean = self.posterior_alpha / (self.posterior_alpha + self.posterior_beta)
        mode = (self.posterior_alpha - 1) / (self.posterior_alpha + self.posterior_beta - 2)
        std_dev = np.sqrt((self.posterior_alpha * self.posterior_beta) /
                        ((self.posterior_alpha + self.posterior_beta)**2 *
                         (self.posterior_alpha + self.posterior_beta + 1)))

        return {
            'mean': mean,
            'mode': mode,
            'std_dev': std_dev,
            'credible_interval': self._calculate_credible_interval()
        }

    def _calculate_credible_interval(self, ci=0.95):
        """Calculate credible interval"""
        lower = beta.ppf((1 - ci)/2, self.posterior_alpha, self.posterior_beta)
        upper = beta.ppf(1 - (1 - ci)/2, self.posterior_alpha, self.posterior_beta)
        return (lower, upper)

# Example Usage
if __name__ == "__main__":
    # Initialize with uniform prior (Beta(1,1))
    bayes = BayesianInference()

    # Update with observed data: 10 successes, 5 failures
    bayes.update(successes=10, failures=5)

    # Get posterior statistics
    stats = bayes.get_posterior_stats()
    print("Posterior Statistics:")
    print(f"Mean: {stats['mean']:.4f}")
    print(f"Mode: {stats['mode']:.4f}")
    print(f"Standard Deviation: {stats['std_dev']:.4f}")
    print(f"95% Credible Interval: ({stats['credible_interval'][0]:.4f}, {stats['credible_interval'][1]:.4f})")

    # Visualize distributions
    bayes.plot_distributions()
```

```python
#12 Perform a Chi-square test for independence between two categorical variables in Python.

import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

def chi_square_test(data, alpha=0.05):
    """
    Perform chi-square test for independence on categorical data.

    Parameters:
    data : array-like, pd.DataFrame, or pd.crosstab
        Contingency table of observed counts
    alpha : float
        Significance level (default: 0.05)

    Returns:
    dict: Test results and statistics
    """
    # Convert data to numpy array if it isn't already
    observed = np.array(data)

    # Perform chi-square test
    chi2, p, dof, expected = chi2_contingency(observed)

    # Determine whether to reject null hypothesis
    reject_null = p < alpha

    # Calculate effect size (Cramer's V)
    n = observed.sum()
    phi = np.sqrt(chi2/n)
    min_dim = min(observed.shape) - 1
    cramers_v = phi / np.sqrt(min_dim)

    return {
        'chi2_statistic': chi2,
        'p_value': p,
        'degrees_of_freedom': dof,
        'expected_counts': expected,
        'reject_null': reject_null,
        'effect_size': cramers_v,
        'significance_level': alpha
    }

def print_results(results):
    """Print formatted test results"""
    print("\nChi-Square Test Results:")
    print("=" * 40)
    print(f"Chi-Square Statistic: {results['chi2_statistic']:.4f}")
    print(f"p-value: {results['p_value']:.4f}")
    print(f"Degrees of Freedom: {results['degrees_of_freedom']}")
    print(f"Significance Level: {results['significance_level']}")
    print(f"Reject Null Hypothesis? {'Yes' if results['reject_null'] else 'No'}")
    print(f"Effect Size (Cramer's V): {results['effect_size']:.4f}")

    print("\nExpected Counts:")
    print(pd.DataFrame(results['expected_counts']))

    # Interpret effect size
    v = results['effect_size']
    if v < 0.1:
        strength = "Negligible"
    elif v < 0.3:
        strength = "Small"
    elif v < 0.5:
        strength = "Medium"
    else:
        strength = "Large"
    print(f"\nEffect Size Interpretation: {strength}")

# Example Usage
if __name__ == "__main__":
    # Create example contingency table
    # Let's test if movie preference is independent of gender
    data = pd.DataFrame({
        'Action': [30, 20],  # Male, Female
        'Comedy': [15, 30],
        'Drama': [10, 25]
    }, index=['Male', 'Female'])

    print("Observed Counts:")
    print(data)

    # Perform test
    results = chi_square_test(data)

    # Print results
    print_results(results)
```

```python
#13 Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.

import numpy as np

def calculate_expected_frequencies(observed):
    """
    Calculates the expected frequencies for a Chi-square test given an observed frequency table.

    Args:
        observed (numpy.ndarray or list of lists): A 2D array or list of lists representing the
                                                   observed frequencies in a contingency table.
                                                   Rows represent one categorical variable,
                                                   and columns represent the other.
                                                   For a goodness-of-fit test, it should be a 1D array or list.

    Returns:
        numpy.ndarray: A NumPy array of the same shape as the observed data, containing the
                       expected frequencies. Returns None if input is invalid.
    """
    observed = np.array(observed)

    if observed.ndim == 1:
        # Goodness-of-fit test: Expected frequencies are often based on a theoretical distribution
        # In this general function, we'll assume equal expected proportions if no other info is given.
        total_observed = np.sum(observed)
        num_categories = len(observed)
        expected = np.array([total_observed / num_categories] * num_categories)
        return expected
    elif observed.ndim == 2:
        # Test of independence or homogeneity:
        row_totals = np.sum(observed, axis=1, keepdims=True)
        col_totals = np.sum(observed, axis=0, keepdims=True)
        grand_total = np.sum(observed)

        expected = (row_totals @ col_totals) / grand_total
        return expected
    else:
        print("Invalid input shape. Observed data should be 1D or 2D.")
        return None

if __name__ == "__main__":
    print("--- Example 1: Goodness-of-Fit Test ---")
    observed_gof = [25, 18, 22, 15, 20]  # Observed frequencies for 5 categories
    expected_gof = calculate_expected_frequencies(observed_gof)

    if expected_gof is not None:
        print("Observed Frequencies:", observed_gof)
        print("Expected Frequencies (assuming equal proportions):", expected_gof)

    print("\n--- Example 2: Test of Independence (Contingency Table) ---")
    observed_contingency = [
        [45, 30, 25],
        [35, 40, 45]
    ]
    expected_contingency = calculate_expected_frequencies(observed_contingency)

    if expected_contingency is not None:
        print("Observed Contingency Table:")
        for row in observed_contingency:
            print(row)
        print("\nExpected Frequencies Table:")
        print(expected_contingency)

    print("\n--- Example 3: Test of Homogeneity ---")
    # The calculation for expected frequencies is the same as for independence
    observed_homogeneity = [
        [60, 40],
        [55, 45],
        [70, 30]
    ]
    expected_homogeneity = calculate_expected_frequencies(observed_homogeneity)

    if expected_homogeneity is not None:
        print("Observed Homogeneity Table:")
        for row in observed_homogeneity:
            print(row)
        print("\nExpected Frequencies Table:")
        print(expected_homogeneity)
```

```python
#14 Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

import numpy as np
from scipy import stats

def goodness_of_fit_test(observed, expected=None, alpha=0.05):
    """
    Performs a Chi-square goodness-of-fit test to compare observed data to an expected distribution.

    Args:
        observed (numpy.ndarray or list): A 1D array or list representing the observed frequencies
                                           for each category.
        expected (numpy.ndarray or list, optional): A 1D array or list representing the expected
                                                    frequencies for each category. If None, assumes
                                                    a uniform distribution (equal expected frequencies).
                                                    Defaults to None.
        alpha (float, optional): The significance level. Defaults to 0.05.

    Returns:
        tuple: A tuple containing the Chi-square statistic and the p-value.
               Returns (None, None) if input is invalid.
    """
    observed = np.array(observed)

    if observed.ndim != 1:
        print("Error: Observed data must be a 1D array or list.")
        return None, None

    k = len(observed)  # Number of categories

    if expected is None:
        total_observed = np.sum(observed)
        expected = np.array([total_observed / k] * k)
    else:
        expected = np.array(expected)
        if expected.ndim != 1 or len(expected) != k:
            print("Error: Expected data must be a 1D array or list with the same length as observed.")
            return None, None
        if np.any(expected <= 0):
            print("Error: Expected frequencies must be greater than zero.")
            return None, None

    if np.any(observed < 0):
        print("Error: Observed frequencies cannot be negative.")
        return None, None

    # Calculate the Chi-square statistic
    chi2_statistic = np.sum((observed - expected)**2 / expected)

    # Calculate the p-value
    degrees_freedom = k - 1
    p_value = 1 - stats.chi2.cdf(chi2_statistic, degrees_freedom)

    return chi2_statistic, p_value

def interpret_goodness_of_fit_test(chi2_statistic, p_value, alpha):
    """
    Interprets the results of the Chi-square goodness-of-fit test.

    Args:
        chi2_statistic (float): The calculated Chi-square statistic.
        p_value (float): The calculated p-value.
        alpha (float): The significance level.
    """
    print("\n--- Chi-square Goodness-of-Fit Test Results ---")
    print(f"Chi-square Statistic: {chi2_statistic:.4f}")
    print(f"P-Value: {p_value:.4f}")
    print(f"Significance Level (alpha): {alpha}")

    if p_value < alpha:
        print("\nConclusion:")
        print(f"The p-value ({p_value:.4f}) is less than the significance level ({alpha}).")
        print("We reject the null hypothesis.")
        print("There is statistically significant evidence that the observed frequencies do not fit the expected distribution.")
    else:
        print("\nConclusion:")
        print(f"The p-value ({p_value:.4f}) is greater than or equal to the significance level ({alpha}).")
        print("We fail to reject the null hypothesis.")
        print("There is not enough statistically significant evidence to conclude that the observed frequencies are significantly different from the expected distribution.")

if __name__ == "__main__":
    # Example 1: Testing for a fair die (uniform distribution)
    observed_die = [12, 15, 10, 18, 16, 9]  # Outcomes of 80 rolls
    alpha = 0.05

    chi2_die, p_value_die = goodness_of_fit_test(observed_die, alpha=alpha)

    if chi2_die is not None:
        interpret_goodness_of_fit_test(chi2_die, p_value_die, alpha)

    print("\n--- Example 2: Comparing to a specific expected distribution ---")
    observed_colors = [45, 30, 25]  # Observed counts of red, blue, green candies
    expected_proportions = [0.5, 0.3, 0.2]
    total_observed_colors = np.sum(observed_colors)
    expected_colors = [total_observed_colors * p for p in expected_proportions]

    chi2_colors, p_value_colors = goodness_of_fit_test(observed_colors, expected_colors, alpha=alpha)

    if chi2_colors is not None:
        print("\nObserved Frequencies (Colors):", observed_colors)
        print("Expected Frequencies (Based on Proportions):", expected_colors)
        interpret_goodness_of_fit_test(chi2_colors, p_value_colors, alpha)

    print("\n--- Example 3: Invalid Input (2D array) ---")
    invalid_observed = [[10, 20], [30, 40]]
    chi2_invalid, p_value_invalid = goodness_of_fit_test(invalid_observed, alpha=alpha)
    print("Chi-square for invalid input:", chi2_invalid, p_value_invalid)

    print("\n--- Example 4: Invalid Input (Expected length mismatch) ---")
    mismatched_observed = [10, 20, 30]
    mismatched_expected = [30, 70]
    chi2_mismatch, p_value_mismatch = goodness_of_fit_test(mismatched_observed, mismatched_expected, alpha=alpha)
    print("Chi-square for mismatched input:", chi2_mismatch, p_value_mismatch)
```

```python
#15 Create a Python script to simulate and visualize the Chi-square distribution and discuss its characteristics.

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def simulate_chi2_distribution(degrees_freedom, num_simulations=10000):
    """
    Simulates the Chi-square distribution by summing the squares of
    standard normal random variables.

    Args:
        degrees_freedom (int): The number of degrees of freedom.
        num_simulations (int, optional): The number of simulations to run.
                                         Defaults to 10000.

    Returns:
        numpy.ndarray: An array of simulated Chi-square values.
    """
    if degrees_freedom <= 0 or not isinstance(degrees_freedom, int):
        raise ValueError("Degrees of freedom must be a positive integer.")

    chi2_values = np.sum(np.random.randn(num_simulations, degrees_freedom)**2, axis=1)
    return chi2_values

def visualize_chi2_distribution(degrees_freedom, simulated_data=None, num_bins=50):
    """
    Visualizes the Chi-square distribution using both simulated and theoretical data.

    Args:
        degrees_freedom (int): The number of degrees of freedom.
        simulated_data (numpy.ndarray, optional): Array of simulated Chi-square values.
                                                  If None, simulation will be performed.
                                                  Defaults to None.
        num_bins (int, optional): The number of bins for the histogram. Defaults to 50.
    """
    if simulated_data is None:
        simulated_data = simulate_chi2_distribution(degrees_freedom)

    plt.figure(figsize=(10, 6))

    # Plot the histogram of simulated data
    plt.hist(simulated_data, bins=num_bins, density=True, alpha=0.6, color='skyblue',
             label=f'Simulated Chi-square (df={degrees_freedom})')

    # Plot the theoretical Chi-square PDF
    x = np.linspace(stats.chi2.ppf(0.0001, degrees_freedom),
                    stats.chi2.ppf(0.9999, degrees_freedom), 200)
    plt.plot(x, stats.chi2.pdf(x, degrees_freedom), 'r-', lw=2, label=f'Theoretical Chi-square PDF (df={degrees_freedom})')

    plt.title(f'Simulation and Visualization of Chi-square Distribution (df={degrees_freedom})')
    plt.xlabel('Chi-square Value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True, alpha=0.5)
    plt.show()

def discuss_chi2_characteristics():
    """Discusses the key characteristics of the Chi-square distribution."""
    print("\n--- Characteristics of the Chi-square Distribution ---")
    print("1. Non-negative:")
    print("   The Chi-square distribution only takes non-negative values (greater than or equal to zero).")
    print("   This is because it is the sum of squared values, which are always non-negative.")

    print("\n2. Defined by Degrees of Freedom (df):")
    print("   The shape of the Chi-square distribution is determined by its degrees of freedom (df).")
    print("   The df is a positive integer that typically relates to the number of independent")
    print("   pieces of information used to calculate a statistic.")

    print("\n3. Right-Skewed:")
    print("   The Chi-square distribution is positively skewed (right-skewed). The degree of skewness")
    print("   decreases as the degrees of freedom increase. For larger df, the distribution becomes")
    print("   more symmetric and approaches a normal distribution (due to the Central Limit Theorem).")

    print("\n4. Mean and Variance:")
    print("   - The mean of a Chi-square distribution with df = k is equal to k.")
    print("   - The variance of a Chi-square distribution with df = k is equal to 2k.")

    print("\n5. Additive Property:")
    print("   If independent random variables X1, X2, ..., Xn follow Chi-square distributions with")
    print("   df1, df2, ..., dfn respectively, then their sum (X1 + X2 + ... + Xn) also follows a")
    print("   Chi-square distribution with degrees of freedom equal to the sum of the individual")
    print("   degrees of freedom (df1 + df2 + ... + dfn).")

    print("\n6. Relationship to Normal Distribution:")
    print("   A Chi-square distribution with k degrees of freedom is the distribution of the sum of")
    print("   the squares of k independent standard normal random variables (Z ~ N(0, 1)).")

    print("\n7. Applications in Statistics:")
    print("   The Chi-square distribution is widely used in various statistical tests, including:")
    print("   - Goodness-of-fit tests (to see if observed data fits an expected distribution).")
    print("   - Tests of independence (to see if two categorical variables are independent).")
    print("   - Tests of homogeneity (to see if the distribution of a categorical variable is the same across different groups).")
    print("   - Confidence interval estimation for variance of a normally distributed population.")

if __name__ == "__main__":
    degrees_of_freedom_values = [1, 3, 5, 10]

    for df in degrees_of_freedom_values:
        simulated_data = simulate_chi2_distribution(df)
        visualize_chi2_distribution(df, simulated_data)

    discuss_chi2_characteristics()
```

```python
#16 Implement an F-test using Python to compare the variances of two random samples.

import numpy as np
from scipy import stats

def f_test_variance(sample1, sample2, alpha=0.05, alternative='two-sided'):
    """
    Performs an F-test to compare the variances of two independent random samples.

    Args:
        sample1 (numpy.ndarray or list): The data for the first sample.
        sample2 (numpy.ndarray or list): The data for the second sample.
        alpha (float, optional): The significance level. Defaults to 0.05.
        alternative (str, optional): The type of alternative hypothesis.
                                     Can be 'two-sided', 'less', or 'greater'.
                                     'two-sided': Variances are not equal (default).
                                     'less': Variance of sample1 is less than variance of sample2.
                                     'greater': Variance of sample1 is greater than variance of sample2.

    Returns:
        tuple: A tuple containing the F-statistic and the p-value.
               Returns (None, None) if input is invalid.
    """
    sample1 = np.array(sample1)
    sample2 = np.array(sample2)

    if sample1.ndim != 1 or sample2.ndim != 1:
        print("Error: Input samples must be 1D arrays or lists.")
        return None, None

    if len(sample1) < 2 or len(sample2) < 2:
        print("Error: Each sample must have at least 2 data points to calculate variance.")
        return None, None

    var1 = np.var(sample1, ddof=1)  # Sample variance (unbiased estimator)
    var2 = np.var(sample2, ddof=1)  # Sample variance (unbiased estimator)

    if var2 == 0:
        print("Error: Variance of the second sample is zero. F-test is undefined.")
        return None, None
    elif var1 == 0 and alternative == 'two-sided':
        f_statistic = 0
    elif var1 == 0 and alternative == 'greater':
        f_statistic = np.inf
    elif var1 == 0 and alternative == 'less':
        f_statistic = 0
    else:
        f_statistic = var1 / var2

    dfn = len(sample1) - 1  # Degrees of freedom for numerator
    dfd = len(sample2) - 1  # Degrees of freedom for denominator

    if alternative == 'two-sided':
        p_value = 2 * min(stats.f.cdf(f_statistic, dfn, dfd), 1 - stats.f.cdf(f_statistic, dfn, dfd))
    elif alternative == 'less':
        p_value = stats.f.cdf(f_statistic, dfn, dfd)
    elif alternative == 'greater':
        p_value = 1 - stats.f.cdf(f_statistic, dfn, dfd)
    else:
        raise ValueError("Invalid alternative hypothesis. Choose 'two-sided', 'less', or 'greater'.")

    return f_statistic, p_value

def interpret_f_test_variance(f_statistic, p_value, alpha, alternative, var1, var2):
    """
    Interprets the results of the F-test for variances.

    Args:
        f_statistic (float): The calculated F-statistic.
        p_value (float): The calculated p-value.
        alpha (float): The significance level.
        alternative (str): The type of alternative hypothesis.
        var1 (float): Variance of the first sample.
        var2 (float): Variance of the second sample.
    """
    print("\n--- F-test for Equality of Variances ---")
    print(f"F-Statistic: {f_statistic:.4f}")
    print(f"P-Value: {p_value:.4f}")
    print(f"Significance Level (alpha): {alpha}")
    print(f"Alternative Hypothesis: {alternative}")
    print(f"Variance of Sample 1: {var1:.4f}")
    print(f"Variance of Sample 2: {var2:.4f}")

    if p_value < alpha:
        print("\nConclusion:")
        print(f"The p-value ({p_value:.4f}) is less than the significance level ({alpha}).")
        print("We reject the null hypothesis that the variances of the two populations are equal.")
        if alternative == 'two-sided':
            print("There is statistically significant evidence that the variances of the two populations are different.")
        elif alternative == 'less':
            print("There is statistically significant evidence that the variance of the first population is less than the variance of the second population.")
        elif alternative == 'greater':
            print("There is statistically significant evidence that the variance of the first population is greater than the variance of the second population.")
    else:
        print("\nConclusion:")
        print(f"The p-value ({p_value:.4f}) is greater than or equal to the significance level ({alpha}).")
        print("We fail to reject the null hypothesis that the variances of the two populations are equal.")
        if alternative == 'two-sided':
            print("There is not enough statistically significant evidence to conclude that the variances of the two populations are different.")
        elif alternative == 'less':
            print("There is not enough statistically significant evidence to conclude that the variance of the first population is less than the variance of the second population.")
        elif alternative == 'greater':
            print("There is not enough statistically significant evidence to conclude that the variance of the first population is greater than the variance of the second population.")

if __name__ == "__main__":
    # Example Usage:
    np.random.seed(42)
    sample_group1 = np.random.normal(loc=5, scale=2, size=50)
    sample_group2 = np.random.normal(loc=5.5, scale=2.5, size=60)
    alpha = 0.05
    alternative_hypothesis = 'two-sided'  # Try 'less' or 'greater'

    # Perform the F-test
    f_statistic, p_value = f_test_variance(sample_group1, sample_group2, alpha, alternative_hypothesis)

    # Interpret the results
    if f_statistic is not None:
        var1 = np.var(sample_group1, ddof=1)
        var2 = np.var(sample_group2, ddof=1)
        interpret_f_test_variance(f_statistic, p_value, alpha, alternative_hypothesis, var1, var2)
```

```python
#17 Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results.

import numpy as np
from scipy import stats
import pandas as pd
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

def perform_anova(data, group_column, value_column, alpha=0.05):
    """
    Performs a one-way ANOVA test on the provided data.

    Args:
        data (pandas.DataFrame): DataFrame containing the data.
        group_column (str): Name of the column representing the groups.
        value_column (str): Name of the column containing the values to compare.
        alpha (float, optional): The significance level. Defaults to 0.05.

    Returns:
        pandas.DataFrame: ANOVA table.
    """
    formula = f'{value_column} ~ C({group_column})'
    model = ols(formula, data=data).fit()
    anova_table = anova_lm(model)
    return anova_table

def interpret_anova_results(anova_table, alpha):
    """
    Interprets the results of the ANOVA test.

    Args:
        anova_table (pandas.DataFrame): ANOVA table from perform_anova.
        alpha (float): The significance level.
    """
    print("\n--- ANOVA Test Results ---")
    print(anova_table)

    p_value = anova_table['PR(>F)'][0]

    print(f"\nSignificance Level (alpha): {alpha}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print("\nConclusion:")
        print(f"The p-value ({p_value:.4f}) is less than the significance level ({alpha}).")
        print("We reject the null hypothesis.")
        print("There is statistically significant evidence that the means of at least one group are different.")
    else:
        print("\nConclusion:")
        print(f"The p-value ({p_value:.4f}) is greater than or equal to the significance level ({alpha}).")
        print("We fail to reject the null hypothesis.")
        print("There is not enough statistically significant evidence to conclude that the means of the groups are different.")
        print("This does not mean the means are equal, only that we don't have sufficient evidence to say they are different.")

if __name__ == "__main__":
    # --- Simulate Data for Multiple Groups ---
    np.random.seed(42)
    group1 = np.random.normal(loc=50, scale=10, size=30)
    group2 = np.random.normal(loc=55, scale=12, size=35)
    group3 = np.random.normal(loc=48, scale=9, size=40)

    # Create a Pandas DataFrame
    data = pd.DataFrame({
        'Group': ['A'] * len(group1) + ['B'] * len(group2) + ['C'] * len(group3),
        'Value': np.concatenate([group1, group2, group3])
    })

    print("Sample Data:")
    print(data.head())

    # --- Perform ANOVA Test ---
    alpha = 0.05
    anova_results = perform_anova(data, group_column='Group', value_column='Value', alpha=alpha)

    # --- Interpret the Results ---
    interpret_anova_results(anova_results, alpha)

    print("\n--- Important Notes on ANOVA ---")
    print("1. Assumptions:")
    print("   - Independence of observations within and between groups.")
    print("   - Normality of the distribution of the response variable for each group.")
    print("   - Homogeneity of variances (equal variances) across the groups.")
    print("   Violations of these assumptions can affect the validity of the ANOVA results.")
    print("\n2. Post-Hoc Tests:")
    print("   If the ANOVA test rejects the null hypothesis (i.e., we find a significant difference"),
    print("   it only tells us that at least one group mean is different. To determine which specific")
    print("   group(s) differ from each other, post-hoc tests (e.g., Tukey's HSD, Bonferroni) are typically performed.")
    print("\n3. One-Way ANOVA:")
    print("   The example here demonstrates a one-way ANOVA, which examines the effect of one")
    print("   categorical independent variable (the 'Group' column) on one continuous dependent variable (the 'Value' column).")
    print("\n4. Two-Way ANOVA and Higher:")
    print("   For examining the effects of two or more independent variables, or the interaction between them,")
    print("   you would use two-way ANOVA or higher-order ANOVA tests.")
```

```python
#19 Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

import numpy as np
import pandas as pd
from scipy import stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.diagnostic import het_breuschpagan
import matplotlib.pyplot as plt
import seaborn as sns

def check_anova_assumptions(data, group_column, value_column, alpha=0.05):
    """
    Checks the assumptions of one-way ANOVA: normality of residuals,
    independence of observations, and homogeneity of variances.

    Args:
        data (pandas.DataFrame): DataFrame containing the data.
        group_column (str): Name of the column representing the groups.
        value_column (str): Name of the column containing the values to compare.
        alpha (float, optional): The significance level for hypothesis tests. Defaults to 0.05.

    Returns:
        pandas.DataFrame: DataFrame containing the results of the assumption checks.
    """
    results = {}

    # 1. Normality of Residuals
    formula = f'{value_column} ~ C({group_column})'
    model = ols(formula, data=data).fit()
    residuals = model.resid

    # Shapiro-Wilk test for normality
    shapiro_test = stats.shapiro(residuals)
    results['Shapiro-Wilk Test (Normality)'] = {
        'Statistic': shapiro_test.statistic,
        'P-value': shapiro_test.pvalue,
        'Interpretation': f'P-value = {shapiro_test.pvalue:.3f}. Residuals appear normally distributed'
                          if shapiro_test.pvalue > alpha
                          else f'P-value = {shapiro_test.pvalue:.3f}. Residuals do not appear normally distributed'
    }

    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    sns.histplot(residuals, kde=True)
    plt.title('Histogram of Residuals')

    plt.subplot(1, 2, 2)
    sm.qqplot(residuals, line='s', fit=True)
    plt.title('QQ-Plot of Residuals')
    plt.tight_layout()
    plt.show()

    # 2. Independence of Observations
    # This assumption is usually checked based on the experimental design
    # and is harder to verify statistically from the data itself.
    # We will provide a note about this.
    results['Independence of Observations'] = 'Checked based on experimental design. Assumed independent if data collection process ensured no dependency between observations.'

    # 3. Homogeneity of Variances (Equal Variances)
    # Levene's test
    groups = data[group_column].unique()
    group_data = [data[data[group_column] == group][value_column] for group in groups]
    levene_test = stats.levene(*group_data)
    results['Levene\'s Test (Homogeneity of Variances)'] = {
        'Statistic': levene_test.statistic,
        'P-value': levene_test.pvalue,
        'Interpretation': f'P-value = {levene_test.pvalue:.3f}. Variances appear homogeneous'
                          if levene_test.pvalue > alpha
                          else f'P-value = {levene_test.pvalue:.3f}. Variances do not appear homogeneous'
    }

    # Breusch-Pagan test (another test for heteroscedasticity)
    bp_test = het_breuschpagan(residuals, model.exog)
    results['Breusch-Pagan Test (Homogeneity of Variances)'] = {
        'LM Statistic': bp_test[0],
        'LM P-value': bp_test[1],
        'F Statistic': bp_test[2],
        'F P-value': bp_test[3],
        'Interpretation (LM)': f'P-value (LM) = {bp_test[1]:.3f}. Variances appear homogeneous'
                               if bp_test[1] > alpha
                               else f'P-value (LM) = {bp_test[1]:.3f}. Variances do not appear homogeneous',
        'Interpretation (F)': f'P-value (F) = {bp_test[3]:.3f}. Variances appear homogeneous'
                              if bp_test[3] > alpha
                              else f'P-value (F) = {bp_test[3]:.3f}. Variances do not appear homogeneous'
    }

    return pd.DataFrame.from_dict(results, orient='index')

if __name__ == "__main__":
    # --- Simulate Data for Multiple Groups (Satisfying Assumptions) ---
    np.random.seed(42)
    group_a = np.random.normal(loc=50, scale=10, size=30)
    group_b = np.random.normal(loc=55, scale=10, size=35)
    group_c = np.random.normal(loc=48, scale=10, size=40)

    data_good = pd.DataFrame({
        'Group': ['A'] * len(group_a) + ['B'] * len(group_b) + ['C'] * len(group_c),
        'Value': np.concatenate([group_a, group_b, group_c])
    })

    print("--- Checking ANOVA Assumptions for Data Satisfying Assumptions ---")
    assumption_results_good = check_anova_assumptions(data_good, group_column='Group', value_column='Value')
    print(assumption_results_good)

    print("\n--- Notes on Assumption Checking ---")
    print("1. Normality: The Shapiro-Wilk test assesses if the residuals are normally distributed. A p-value greater than alpha suggests normality. The histogram and QQ-plot provide visual checks.")
    print("2. Independence: Crucial and usually determined by the experimental design. Assume independence if data collection was random and without dependence between observations.")
    print("3. Homogeneity of Variances: Levene's and Breusch-Pagan tests check for equal variances across groups. P-values greater than alpha suggest homogeneity.")
    print("\nIt's important to address violations of these assumptions before relying heavily on ANOVA results. Data transformations or non-parametric tests might be needed.")

    print("\n--- Example of Data Violating Homogeneity of Variances ---")
    np.random.seed(123)
    group_x = np.random.normal(loc=20, scale=5, size=30)
    group_y = np.random.normal(loc=25, scale=15, size=35)
    group_z = np.random.normal(loc=22, scale=8, size=40)
    heterogeneous_data = pd.DataFrame({
        'Group': ['X'] * len(group_x) + ['Y'] * len(group_y) + ['Z'] * len(group_z),
        'Value': np.concatenate([group_x, group_y, group_z])
    })
    print("\n--- Checking Assumptions for Data Violating Homogeneity of Variances ---")
    hetero_assumption_results = check_anova_assumptions(heterogeneous_data, group_column='Group', value_column='Value')
    print(hetero_assumption_results)

    print("\n--- Example of Data Potentially Violating Normality ---")
    np.random.seed(789)
    group_p = np.concatenate([np.random.normal(loc=30, scale=5, size=25), np.random.exponential(scale=10, size=15)])
    group_q = np.concatenate([np.random.normal(loc=35, scale=6, size=28), np.random.poisson(lam=5, size=12)])
    non_normal_data = pd.DataFrame({
        'Group': ['P'] * len(group_p) + ['Q'] * len(group_q),
        'Value': np.concatenate([group_p, group_q])
    })
    print("\n--- Checking Assumptions for Data Potentially Violating Normality ---")
    non_normal_assumption_results = check_anova_assumptions(non_normal_data, group_column='Group', value_column='Value')
    print(non_normal_assumption_results)
```

```python
#20 Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
from statsmodels.graphics.factorplots import interaction_plot

# Generate sample data with interaction effect
np.random.seed(42)

# Create experimental factors
levels_a = ['A1', 'A2']
levels_b = ['B1', 'B2']
n = 20  # Observations per group

data = []
for a in levels_a:
    for b in levels_b:
        # Create interaction effect
        if a == 'A1' and b == 'B1':
            mu = 50
        elif a == 'A1' and b == 'B2':
            mu = 65
        elif a == 'A2' and b == 'B1':
            mu = 70
        else:
            mu = 55  # This creates the interaction effect
        group_data = np.random.normal(mu, 10, n)
        for value in group_data:
            data.append([a, b, value])

df = pd.DataFrame(data, columns=['FactorA', 'FactorB', 'Value'])

# Perform two-way ANOVA with interaction
model = ols('Value ~ C(FactorA) + C(FactorB) + C(FactorA):C(FactorB)', data=df).fit()
anova_table = anova_lm(model, typ=2)

print("Two-Way ANOVA Table:")
print("="*50)
print(anova_table)
print("\nKey:")
print("- 'PR(>F)' shows p-values")
print("- C(FactorA):C(FactorB) represents the interaction effect")

# Visualize interaction effects
plt.figure(figsize=(15, 6))

# Interaction plot 1
plt.subplot(1, 2, 1)
interaction_plot(
    x=df['FactorA'],
    trace=df['FactorB'],
    response=df['Value'],
    colors=['#4c72b0', '#dd8452'],
    ms=10
)
plt.title('Interaction Plot: Factor A vs Factor B')
plt.ylabel('Mean Value')

# Interaction plot 2 (alternative visualization)
plt.subplot(1, 2, 2)
sns.pointplot(
    x='FactorA',
    y='Value',
    hue='FactorB',
    data=df,
    ci=95,
    palette='Set2',
    dodge=True
)
plt.title('Mean Values with 95% Confidence Intervals')
plt.ylabel('Value')
plt.xlabel('Factor A')

plt.tight_layout()
plt.show()

# Heatmap of interaction effects
interaction_effects = df.groupby(['FactorA', 'FactorB'])['Value'].mean().unstack()
plt.figure(figsize=(8, 6))
sns.heatmap(
    interaction_effects,
    annot=True,
    cmap='coolwarm',
    fmt=".1f",
    linewidths=.5
)
plt.title('Interaction Effect Heatmap')
plt.ylabel('Factor A')
plt.xlabel('Factor B')
plt.show()

# Residual analysis
fig = plt.figure(figsize=(12, 5))
ax1 = fig.add_subplot(1, 2, 1)
sm.qqplot(model.resid, line='s', ax=ax1)
ax1.set_title('Q-Q Plot of Residuals')

ax2 = fig.add_subplot(1, 2, 2)
ax2.scatter(model.fittedvalues, model.resid)
ax2.axhline(y=0, color='r', linestyle='--')
ax2.set_title('Residuals vs Fitted Values')
ax2.set_xlabel('Fitted Values')
ax2.set_ylabel('Residuals')

plt.tight_layout()
plt.show()
```

```python
#21 Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing.

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def visualize_f_distribution(dfn_values, dfd_values, x_max=10, num_points=500):
    """
    Visualizes the F-distribution for different degrees of freedom.

    Args:
        dfn_values (list of int): List of numerator degrees of freedom.
        dfd_values (list of int): List of denominator degrees of freedom.
        x_max (int, optional): Maximum value on the x-axis for plotting. Defaults to 10.
        num_points (int, optional): Number of points to generate for the plot. Defaults to 500.
    """
    x = np.linspace(0.01, x_max, num_points)
    colors = plt.cm.viridis(np.linspace(0, 1, len(dfn_values) * len(dfd_values)))
    color_index = 0

    plt.figure(figsize=(12, 8))

    for dfn in dfn_values:
        for dfd in dfd_values:
            pdf = stats.f.pdf(x, dfn, dfd)
            plt.plot(x, pdf, color=colors[color_index],
                     label=f'dfn={dfn}, dfd={dfd}')
            color_index += 1

    plt.title('Visualization of the F-Distribution')
    plt.xlabel('F-statistic Value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True, alpha=0.5)
    plt.xlim(0, x_max)
    plt.ylim(0, max(stats.f.pdf(0.5, max(dfn_values), max(dfd_values))))
    plt.show()

def discuss_f_distribution_in_hypothesis_testing():
    """Discusses the use of the F-distribution in hypothesis testing."""
    print("\n--- The F-Distribution and Hypothesis Testing ---")
    print("\n1. Definition and Characteristics:")
    print("   - The F-distribution arises as the ratio of two independent chi-square distributed random variables,")
    print("     each divided by their respective degrees of freedom.")
    print("   - It is characterized by two parameters: degrees of freedom for the numerator (dfn) and degrees of")
    print("     freedom for the denominator (dfd).")
    print("   - The F-distribution is continuous, non-negative, and typically right-skewed. Its shape depends heavily")
    print("     on the values of dfn and dfd.")
    print("   - As the degrees of freedom increase, the F-distribution approaches a normal distribution.")

    print("\n2. Use in Hypothesis Testing:")
    print("   The F-distribution is the foundation for many important statistical tests that compare variances or")
    print("   analyze the variance within and between groups. Key applications include:")

    print("\n   a) Comparing Variances (F-test):")
    print("      - The F-test is used to determine if the variances of two independent populations are significantly")
    print("        different.")
    print("      - The test statistic is the ratio of the two sample variances: F = s₁² / s₂².")
    print("      - Under the null hypothesis that the population variances are equal, this ratio follows an")
    print("        F-distribution with degrees of freedom (n₁-1) for the numerator and (n₂-1) for the denominator,")
    print("        where n₁ and n₂ are the sample sizes.")
    print("      - A large F-statistic (far from 1) provides evidence against the null hypothesis.")
    print("      - The p-value is calculated as the probability of observing an F-statistic as extreme as, or more")
    print("        extreme than, the one calculated, assuming the null hypothesis is true.")

    print("\n   b) Analysis of Variance (ANOVA):")
    print("      - ANOVA is a powerful statistical technique used to compare the means of two or more groups.")
    print("      - It works by partitioning the total variance in the data into different sources of variation.")
    print("      - The F-statistic in ANOVA is calculated as the ratio of the variance between groups (Mean Square Between)")
    print("        to the variance within groups (Mean Square Within or Error Mean Square).")
    print("      - A large F-statistic indicates that the variation between the group means is large relative to the")
    print("        variation within the groups, providing evidence that the group means are not all equal.")
    print("      - The p-value associated with the F-statistic is used to determine the statistical significance of")
    print("        the differences in group means.")
    print("      - One-way ANOVA examines the effect of a single categorical independent variable on a continuous")
    print("        dependent variable. Two-way ANOVA and higher-order ANOVAs can examine the effects of multiple")
    print("        independent variables and their interactions.")

    print("\n3. Interpreting Results:")
    print("   - In both F-tests for variances and ANOVA, a small p-value (typically less than the chosen significance")
    print("     level, e.g., 0.05) leads to the rejection of the null hypothesis.")
    print("   - For the F-test for variances, rejecting the null hypothesis suggests that the population variances are")
    print("     significantly different.")
    print("   - For ANOVA, rejecting the null hypothesis (that all group means are equal) indicates that at least one")
    print("     group mean is significantly different from the others. Post-hoc tests are then used to identify which")
    print("     specific group means differ.")

    print("\n4. Assumptions:")
    print("   - When using the F-distribution in hypothesis testing (especially ANOVA), certain assumptions should be met:")
    print("     - Independence of observations.")
    print("     - Normality of the underlying populations (especially important for smaller sample sizes).")
    print("     - Homogeneity of variances (equal variances across groups for ANOVA).")
    print("   - Violations of these assumptions can affect the validity of the F-test results.")

if __name__ == "__main__":
    # Define different degrees of freedom to visualize
    numerator_dfs = [1, 5, 10]
    denominator_dfs = [5, 10, 30]

    visualize_f_distribution(numerator_dfs, denominator_dfs)
    discuss_f_distribution_in_hypothesis_testing()
```

```python
#23 Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means.

import numpy as np
from scipy import stats

def simulate_normal_data(mean, std_dev, sample_size, seed=None):
    """
    Simulates random data from a normal distribution.

    Args:
        mean (float): The mean of the normal distribution.
        std_dev (float): The standard deviation of the normal distribution.
        sample_size (int): The number of data points to simulate.
        seed (int, optional): Random seed for reproducibility. Defaults to None.

    Returns:
        numpy.ndarray: An array containing the simulated data.
    """
    if seed is not None:
        np.random.seed(seed)
    data = np.random.normal(loc=mean, scale=std_dev, size=sample_size)
    return data

def perform_t_test(sample1, sample2=None, assumed_mean=0, alpha=0.05):
    """
    Performs a one-sample or two-sample t-test to evaluate means.

    Args:
        sample1 (numpy.ndarray): The data for the first sample.
        sample2 (numpy.ndarray, optional): The data for the second sample for a two-sample test. Defaults to None (one-sample test).
        assumed_mean (float, optional): The hypothesized population mean for a one-sample test. Defaults to 0.
        alpha (float, optional): The significance level. Defaults to 0.05.

    Returns:
        tuple: A tuple containing the test statistic and the p-value.
               Returns None if input is invalid.
    """
    if sample1 is None or len(sample1) < 2:
        print("Error: Sample 1 must have at least 2 data points.")
        return None

    if sample2 is None:
        # One-sample t-test
        t_statistic, p_value = stats.ttest_1samp(sample1, assumed_mean)
        print("\n--- One-Sample T-test ---")
        print(f"Sample Mean: {np.mean(sample1):.4f}")
        print(f"Hypothesized Mean: {assumed_mean}")
    elif len(sample2) < 2:
        print("Error: Sample 2 must have at least 2 data points for a two-sample test.")
        return None
    else:
        # Two-sample independent t-test
        t_statistic, p_value = stats.ttest_ind(sample1, sample2)
        print("\n--- Two-Sample Independent T-test ---")
        print(f"Sample 1 Mean: {np.mean(sample1):.4f}")
        print(f"Sample 2 Mean: {np.mean(sample2):.4f}")

    print(f"T-statistic: {t_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print(f"P-value ({p_value:.4f}) is less than alpha ({alpha}).")
        print("Reject the null hypothesis.")
    else:
        print(f"P-value ({p_value:.4f}) is greater than or equal to alpha ({alpha}).")
        print("Fail to reject the null hypothesis.")

    return t_statistic, p_value

if __name__ == "__main__":
    # --- Simulation 1: One-Sample T-test ---
    print("--- Simulation 1: One-Sample T-test ---")
    true_mean_1 = 5
    std_dev_1 = 2
    sample_size_1 = 30
    sample_data_1 = simulate_normal_data(true_mean_1, std_dev_1, sample_size_1, seed=42)

    # Hypothesis test: Is the sample mean significantly different from 4?
    hypothesized_mean_1 = 4
    perform_t_test(sample_data_1, assumed_mean=hypothesized_mean_1)

    # --- Simulation 2: Two-Sample Independent T-test ---
    print("\n--- Simulation 2: Two-Sample Independent T-test ---")
    true_mean_2_group_a = 10
    std_dev_2_group_a = 3
    sample_size_2_group_a = 40
    sample_a = simulate_normal_data(true_mean_2_group_a, std_dev_2_group_a, sample_size_2_group_a, seed=123)

    true_mean_2_group_b = 10.5
    std_dev_2_group_b = 3.5
    sample_size_2_group_b = 45
    sample_b = simulate_normal_data(true_mean_2_group_b, std_dev_2_group_b, sample_size_2_group_b, seed=456)

    # Hypothesis test: Are the means of the two samples significantly different?
    perform_t_test(sample_a, sample_b)

    # --- Simulation 3: One-Sample T-test (Mean is close to hypothesized) ---
    print("\n--- Simulation 3: One-Sample T-test (Mean is close to hypothesized) ---")
    true_mean_3 = 7
    std_dev_3 = 1.5
    sample_size_3 = 50
    sample_data_3 = simulate_normal_data(true_mean_3, std_dev_3, sample_size_3, seed=789)

    # Hypothesis test: Is the sample mean significantly different from 7.1?
    hypothesized_mean_3 = 7.1
    perform_t_test(sample_data_3, assumed_mean=hypothesized_mean_3)
```

```python
#24 Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.

import numpy as np
from scipy import stats

def hypothesis_test_variance_chisquare(sample_data, hypothesized_variance, alpha=0.05, alternative='two-sided'):
    """
    Performs a hypothesis test for population variance using a Chi-square distribution.

    Args:
        sample_data (numpy.ndarray or list): The data for the sample.
        hypothesized_variance (float): The hypothesized population variance (σ₀²).
        alpha (float, optional): The significance level. Defaults to 0.05.
        alternative (str, optional): The type of alternative hypothesis.
                                     Can be 'two-sided', 'less', or 'greater'.
                                     'two-sided': Population variance is not equal to hypothesized variance (default).
                                     'less': Population variance is less than hypothesized variance.
                                     'greater': Population variance is greater than hypothesized variance.

    Returns:
        tuple: A tuple containing the Chi-square statistic and the p-value.
               Returns (None, None) if input is invalid.
    """
    sample_data = np.array(sample_data)
    n = len(sample_data)

    if n < 2:
        print("Error: Sample must have at least 2 data points to calculate variance.")
        return None, None

    sample_variance = np.var(sample_data, ddof=1)  # Unbiased sample variance
    chi2_statistic = (n - 1) * sample_variance / hypothesized_variance
    degrees_of_freedom = n - 1

    if alternative == 'two-sided':
        p_value = 2 * min(stats.chi2.cdf(chi2_statistic, degrees_of_freedom),
                          1 - stats.chi2.cdf(chi2_statistic, degrees_of_freedom))
    elif alternative == 'less':
        p_value = stats.chi2.cdf(chi2_statistic, degrees_of_freedom)
    elif alternative == 'greater':
        p_value = 1 - stats.chi2.cdf(chi2_statistic, degrees_of_freedom)
    else:
        raise ValueError("Invalid alternative hypothesis. Choose 'two-sided', 'less', or 'greater'.")

    print("\n--- Hypothesis Test for Population Variance (Chi-square) ---")
    print(f"Sample Size (n): {n}")
    print(f"Sample Variance (s²): {sample_variance:.4f}")
    print(f"Hypothesized Variance (σ₀²): {hypothesized_variance:.4f}")
    print(f"Chi-square Statistic (χ²): {chi2_statistic:.4f}")
    print(f"Degrees of Freedom (df): {degrees_of_freedom}")
    print(f"Significance Level (alpha): {alpha}")
    print(f"Alternative Hypothesis: {alternative}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print("\nConclusion:")
        print(f"The p-value ({p_value:.4f}) is less than the significance level ({alpha}).")
        print("We reject the null hypothesis that the population variance is equal to the hypothesized variance.")
        if alternative == 'two-sided':
            print(f"There is statistically significant evidence that the population variance is different from {hypothesized_variance:.4f}.")
        elif alternative == 'less':
            print(f"There is statistically significant evidence that the population variance is less than {hypothesized_variance:.4f}.")
        elif alternative == 'greater':
            print(f"There is statistically significant evidence that the population variance is greater than {hypothesized_variance:.4f}.")
    else:
        print("\nConclusion:")
        print(f"The p-value ({p_value:.4f}) is greater than or equal to the significance level ({alpha}).")
        print("We fail to reject the null hypothesis that the population variance is equal to the hypothesized variance.")
        if alternative == 'two-sided':
            print(f"There is not enough statistically significant evidence to conclude that the population variance is different from {hypothesized_variance:.4f}.")
        elif alternative == 'less':
            print(f"There is not enough statistically significant evidence to conclude that the population variance is less than {hypothesized_variance:.4f}.")
        elif alternative == 'greater':
            print(f"There is not enough statistically significant evidence to conclude that the population variance is greater than {hypothesized_variance:.4f}.")

    return chi2_statistic, p_value

if __name__ == "__main__":
    # Example Usage:
    np.random.seed(42)
    population_variance = 25
    sample_size = 30
    sample_data = np.random.normal(loc=10, scale=np.sqrt(population_variance), size=sample_size)

    # Hypothesized variance
    null_hypothesis_variance = 20
    alpha = 0.05

    # Two-sided test: Is the population variance different from 20?
    hypothesis_test_variance_chisquare(sample_data, null_hypothesis_variance, alpha=alpha, alternative='two-sided')

    # One-sided test (less): Is the population variance less than 20?
    hypothesis_test_variance_chisquare(sample_data, null_hypothesis_variance, alpha=alpha, alternative='less')

    # One-sided test (greater): Is the population variance greater than 20?
    hypothesis_test_variance_chisquare(sample_data, null_hypothesis_variance, alpha=alpha, alternative='greater')

    # Another example with a hypothesized variance closer to the true variance
    null_hypothesis_variance_close = 26
    hypothesis_test_variance_chisquare(sample_data, null_hypothesis_variance_close, alpha=alpha, alternative='two-sided')
```

```python
#25 Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.

import numpy as np
import statsmodels.api as sm
from statsmodels.stats.proportion import proportions_ztest

def z_test_proportions(count1, nobs1, count2, nobs2, alpha=0.05, alternative='two-sided'):
    """
    Performs a Z-test for comparing proportions between two datasets or groups.

    Args:
        count1 (int): The number of successes in the first group.
        nobs1 (int): The total number of observations in the first group.
        count2 (int): The number of successes in the second group.
        nobs2 (int): The total number of observations in the second group.
        alpha (float, optional): The significance level. Defaults to 0.05.
        alternative (str, optional): The alternative hypothesis. Can be 'two-sided',
                                     'smaller' (proportion of group1 is smaller than group2),
                                     or 'larger' (proportion of group1 is larger than group2).
                                     Defaults to 'two-sided'.

    Returns:
        tuple: A tuple containing the Z-statistic and the p-value.
    """
    if nobs1 <= 0 or nobs2 <= 0:
        raise ValueError("Number of observations in both groups must be positive.")
    if not (0 <= count1 <= nobs1 and 0 <= count2 <= nobs2):
        raise ValueError("Number of successes must be between 0 and the number of observations for each group.")
    if alternative not in ['two-sided', 'smaller', 'larger']:
        raise ValueError("Invalid alternative hypothesis. Choose 'two-sided', 'smaller', or 'larger'.")

    successes = np.array([count1, count2])
    nobs = np.array([nobs1, nobs2])

    z_statistic, p_value = proportions_ztest(successes, nobs, alternative=alternative)

    print("\n--- Z-test for Comparing Proportions ---")
    print(f"Group 1: Successes = {count1}, Observations = {nobs1}, Proportion = {count1/nobs1:.4f}")
    print(f"Group 2: Successes = {count2}, Observations = {nobs2}, Proportion = {count2/nobs2:.4f}")
    print(f"Significance Level (alpha): {alpha}")
    print(f"Alternative Hypothesis: {alternative}")
    print(f"Z-statistic: {z_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print("\nConclusion:")
        print(f"The p-value ({p_value:.4f}) is less than the significance level ({alpha}).")
        print("Reject the null hypothesis.")
        if alternative == 'two-sided':
            print("There is statistically significant evidence that the proportions of the two groups are different.")
        elif alternative == 'smaller':
            print("There is statistically significant evidence that the proportion of Group 1 is smaller than the proportion of Group 2.")
        elif alternative == 'larger':
            print("There is statistically significant evidence that the proportion of Group 1 is larger than the proportion of Group 2.")
    else:
        print("\nConclusion:")
        print(f"The p-value ({p_value:.4f}) is greater than or equal to the significance level ({alpha}).")
        print("Fail to reject the null hypothesis.")
        if alternative == 'two-sided':
            print("There is not enough statistically significant evidence to conclude that the proportions of the two groups are different.")
        elif alternative == 'smaller':
            print("There is not enough statistically significant evidence to conclude that the proportion of Group 1 is smaller than the proportion of Group 2.")
        elif alternative == 'larger':
            print("There is not enough statistically significant evidence to conclude that the proportion of Group 1 is larger than the proportion of Group 2.")

    return z_statistic, p_value

if __name__ == "__main__":
    # Example 1: Comparing conversion rates of two website designs
    conversions_design_a = 60
    total_visitors_design_a = 500
    conversions_design_b = 85
    total_visitors_design_b = 600
    alpha = 0.05

    z_test_proportions(conversions_design_a, total_visitors_design_a,
                       conversions_design_b, total_visitors_design_b,
                       alpha=alpha, alternative='two-sided')

    print("\n---")

    # Example 2: Testing if the proportion in group 1 is smaller than in group 2
    successes1 = 30
    observations1 = 200
    successes2 = 45
    observations2 = 250
    alpha = 0.01

    z_test_proportions(successes1, observations1, successes2, observations2,
                       alpha=alpha, alternative='smaller')

    print("\n---")

    # Example 3: Testing if the proportion in group 1 is larger than in group 2
    count_group1 = 120
    n_group1 = 300
    count_group2 = 100
    n_group2 = 350
    alpha = 0.10

    z_test_proportions(count_group1, n_group1, count_group2, n_group2,
                       alpha=alpha, alternative='larger')

    print("\n---")

    # Example with potential error (invalid number of successes)
    try:
        z_test_proportions(10, 5, 20, 10, alpha=0.05)
    except ValueError as e:
        print(f"Error: {e}")
```

```python
#26 Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

def f_test_variances(sample1, sample2, alpha=0.05, alternative='two-sided'):
    """
    Performs an F-test to compare the variances of two datasets.

    Args:
        sample1 (numpy.ndarray or list): The data for the first sample.
        sample2 (numpy.ndarray or list): The data for the second sample.
        alpha (float, optional): The significance level. Defaults to 0.05.
        alternative (str, optional): The type of alternative hypothesis.
                                     Can be 'two-sided', 'less' (variance of sample1 < sample2),
                                     or 'greater' (variance of sample1 > sample2).
                                     Defaults to 'two-sided'.

    Returns:
        tuple: A tuple containing the F-statistic and the p-value.
    """
    sample1 = np.array(sample1)
    sample2 = np.array(sample2)
    n1 = len(sample1)
    n2 = len(sample2)

    if n1 < 2 or n2 < 2:
        raise ValueError("Both samples must have at least 2 data points to calculate variance.")

    var1 = np.var(sample1, ddof=1)  # Unbiased sample variance
    var2 = np.var(sample2, ddof=1)  # Unbiased sample variance

    f_statistic = var1 / var2
    dfn = n1 - 1  # Degrees of freedom for numerator
    dfd = n2 - 1  # Degrees of freedom for denominator

    if alternative == 'two-sided':
        p_value = 2 * min(stats.f.cdf(f_statistic, dfn, dfd),
                          1 - stats.f.cdf(f_statistic, dfn, dfd))
    elif alternative == 'less':
        p_value = stats.f.cdf(f_statistic, dfn, dfd)
    elif alternative == 'greater':
        p_value = 1 - stats.f.cdf(f_statistic, dfn, dfd)
    else:
        raise ValueError("Invalid alternative hypothesis. Choose 'two-sided', 'less', or 'greater'.")

    print("\n--- F-test for Comparing Variances ---")
    print(f"Sample 1: Size = {n1}, Variance (s1²) = {var1:.4f}")
    print(f"Sample 2: Size = {n2}, Variance (s2²) = {var2:.4f}")
    print(f"Significance Level (alpha): {alpha}")
    print(f"Alternative Hypothesis: {alternative}")
    print(f"F-statistic (s1²/s2²): {f_statistic:.4f}")
    print(f"Degrees of Freedom (Numerator): {dfn}")
    print(f"Degrees of Freedom (Denominator): {dfd}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print("\nConclusion:")
        print(f"The p-value ({p_value:.4f}) is less than the significance level ({alpha}).")
        print("Reject the null hypothesis that the population variances are equal.")
        if alternative == 'two-sided':
            print("There is statistically significant evidence that the population variances of the two groups are different.")
        elif alternative == 'less':
            print("There is statistically significant evidence that the population variance of Sample 1 is less than that of Sample 2.")
        elif alternative == 'greater':
            print("There is statistically significant evidence that the population variance of Sample 1 is greater than that of Sample 2.")
    else:
        print("\nConclusion:")
        print(f"The p-value ({p_value:.4f}) is greater than or equal to the significance level ({alpha}).")
        print("Fail to reject the null hypothesis that the population variances are equal.")
        if alternative == 'two-sided':
            print("There is not enough statistically significant evidence to conclude that the population variances of the two groups are different.")
        elif alternative == 'less':
            print("There is not enough statistically significant evidence to conclude that the population variance of Sample 1 is less than that of Sample 2.")
        elif alternative == 'greater':
            print("There is not enough statistically significant evidence to conclude that the population variance of Sample 1 is greater than that of Sample 2.")

    # Visualization
    visualize_f_distribution(dfn, dfd, f_statistic, alpha, alternative)

    return f_statistic, p_value

def visualize_f_distribution(dfn, dfd, f_statistic, alpha, alternative):
    """
    Visualizes the F-distribution and the p-value for the F-test.
    """
    x = np.linspace(0.01, max(f_statistic * 2, 5), 500)
    pdf = stats.f.pdf(x, dfn, dfd)
    critical_value_left = stats.f.ppf(alpha / 2, dfn, dfd)
    critical_value_right = stats.f.ppf(1 - alpha / 2, dfn, dfd)

    plt.figure(figsize=(10, 6))
    plt.plot(x, pdf, 'b-', label=f'F({dfn}, {dfd}) Distribution')
    plt.axvline(f_statistic, color='r', linestyle='--', label=f'Observed F = {f_statistic:.2f}')

    if alternative == 'two-sided':
        plt.axvline(critical_value_left, color='g', linestyle=':', label=f'Critical Values (α/2 = {alpha/2:.2f})')
        plt.axvline(critical_value_right, color='g', linestyle=':')
        p_value_region_left = np.linspace(0.01, critical_value_left, 100)
        plt.fill_between(p_value_region_left, stats.f.pdf(p_value_region_left, dfn, dfd), color='lightcoral', alpha=0.5, label='P-value Regions')
        p_value_region_right = np.linspace(critical_value_right, max(f_statistic * 2, 5), 100)
        plt.fill_between(p_value_region_right, stats.f.pdf(p_value_region_right, dfn, dfd), color='lightcoral', alpha=0.5)
    elif alternative == 'less':
        plt.axvline(stats.f.ppf(alpha, dfn, dfd), color='g', linestyle=':', label=f'Critical Value (α = {alpha:.2f})')
        p_value_region = np.linspace(0.01, f_statistic, 100)
        plt.fill_between(p_value_region, stats.f.pdf(p_value_region, dfn, dfd), color='lightcoral', alpha=0.5, label='P-value Region')
    elif alternative == 'greater':
        plt.axvline(stats.f.ppf(1 - alpha, dfn, dfd), color='g', linestyle=':', label=f'Critical Value (α = {alpha:.2f})')
        p_value_region = np.linspace(f_statistic, max(f_statistic * 2, 5), 100)
        plt.fill_between(p_value_region, stats.f.pdf(p_value_region, dfn, dfd), color='lightcoral', alpha=0.5, label='P-value Region')

    plt.title('F-Distribution and Hypothesis Test')
    plt.xlabel('F-statistic Value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True, alpha=0.5)
    plt.show()

if __name__ == "__main__":
    # Example 1: Two-sided test
    np.random.seed(42)
    group1 = np.random.normal(loc=0, scale=1, size=50)
    group2 = np.random.normal(loc=0, scale=1.5, size=60)
    f_test_variances(group1, group2, alpha=0.05, alternative='two-sided')

    print("\n---")

    # Example 2: One-sided test (variance of group1 < group2)
    np.random.seed(123)
    data_a = np.random.normal(loc=10, scale=2, size=40)
    data_b = np.random.normal(loc=10, scale=3, size=45)
    f_test_variances(data_a, data_b, alpha=0.01, alternative='less')

    print("\n---")

    # Example 3: One-sided test (variance of group1 > group2)
    np.random.seed(789)
    set_x = np.random.normal(loc=5, scale=2.5, size=35)
    set_y = np.random.normal(loc=5, scale=1, size=30)
    f_test_variances(set_x, set_y, alpha=0.10, alternative='greater')

    print("\n---")

    # Example with potential error (small sample size)
    try:
        small_group1 = [1, 2]
        small_group2 = [3]
        f_test_variances(small_group1, small_group2)
    except ValueError as e:
        print(f"Error: {e}")
```

```python
#27 Perform a Chi-square test for goodness of fit with simulated data and analyze the results.

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

def chi_square_goodness_of_fit(observed_frequencies, expected_probabilities, alpha=0.05):
    """
    Performs a Chi-square goodness of fit test.

    Args:
        observed_frequencies (list or numpy.ndarray): The observed frequencies for each category.
        expected_probabilities (list or numpy.ndarray): The expected probabilities for each category.
        alpha (float, optional): The significance level. Defaults to 0.05.

    Returns:
        tuple: A tuple containing the Chi-square statistic and the p-value.
    """
    observed = np.array(observed_frequencies)
    expected_prob = np.array(expected_probabilities)
    total_observed = np.sum(observed)
    expected = total_observed * expected_prob

    if np.any(expected == 0):
        raise ValueError("Expected frequencies cannot be zero.")

    chi2_statistic, p_value = stats.chisquare(observed, f_exp=expected)
    degrees_of_freedom = len(observed) - 1

    print("\n--- Chi-square Goodness of Fit Test ---")
    print(f"Observed Frequencies: {observed}")
    print(f"Expected Probabilities: {expected_prob}")
    print(f"Expected Frequencies: {expected.round(2)}")
    print(f"Total Observed: {total_observed}")
    print(f"Chi-square Statistic (χ²): {chi2_statistic:.4f}")
    print(f"Degrees of Freedom (df): {degrees_of_freedom}")
    print(f"Significance Level (alpha): {alpha}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print("\nConclusion:")
        print(f"The p-value ({p_value:.4f}) is less than the significance level ({alpha}).")
        print("Reject the null hypothesis.")
        print("There is statistically significant evidence that the observed frequencies do not fit the expected distribution.")
    else:
        print("\nConclusion:")
        print(f"The p-value ({p_value:.4f}) is greater than or equal to the significance level ({alpha}).")
        print("Fail to reject the null hypothesis.")
        print("There is not enough statistically significant evidence to conclude that the observed frequencies do not fit the expected distribution.")

    # Visualization
    categories = [f'Category {i+1}' for i in range(len(observed))]
    visualize_goodness_of_fit(categories, observed, expected)

    return chi2_statistic, p_value

def visualize_goodness_of_fit(categories, observed, expected):
    """
    Visualizes the observed and expected frequencies.
    """
    x = np.arange(len(categories))
    width = 0.35

    fig, ax = plt.subplots(figsize=(10, 6))
    rects1 = ax.bar(x - width/2, observed, width, label='Observed Frequencies')
    rects2 = ax.bar(x + width/2, expected, width, label='Expected Frequencies')

    ax.set_ylabel('Frequencies')
    ax.set_xlabel('Categories')
    ax.set_title('Observed vs. Expected Frequencies (Chi-square Goodness of Fit)')
    ax.set_xticks(x)
    ax.set_xticklabels(categories)
    ax.legend()

    ax.bar_label(rects1, padding=3)
    ax.bar_label(rects2, padding=3)

    fig.tight_layout()
    plt.show()

if __name__ == "__main__":
    # --- Simulate Data for a Fair Die Roll ---
    np.random.seed(42)
    num_rolls = 300
    outcomes = np.random.randint(1, 7, num_rolls)  # Simulate 300 rolls of a 6-sided die
    observed_counts_fair = np.bincount(outcomes)[1:] # Count occurrences of each face (1 to 6)
    expected_probabilities_fair = [1/6] * 6

    print("--- Testing Goodness of Fit for a Fair Die ---")
    chi_square_goodness_of_fit(observed_counts_fair, expected_probabilities_fair)

    print("\n---")

    # --- Simulate Data for a Potentially Biased Die Roll ---
    np.random.seed(123)
    biased_probabilities = [0.1, 0.15, 0.2, 0.18, 0.22, 0.15] # Example of biased probabilities
    num_rolls_biased = 300
    biased_outcomes = np.random.choice(np.arange(1, 7), size=num_rolls_biased, p=biased_probabilities)
    observed_counts_biased = np.bincount(biased_outcomes)[1:]
    expected_probabilities_biased_null = [1/6] * 6 # Null hypothesis: fair die

    print("--- Testing Goodness of Fit for a Potentially Biased Die (against fair die expectation) ---")
    chi_square_goodness_of_fit(observed_counts_biased, expected_probabilities_biased_null)

    print("\n---")

    # --- Testing Goodness of Fit for the Potentially Biased Die (against its true probabilities) ---
    print("--- Testing Goodness of Fit for a Potentially Biased Die (against true probabilities) ---")
    chi_square_goodness_of_fit(observed_counts_biased, biased_probabilities)
```
